# Visualization

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Popularity'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Song Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of Tempo vs Loudness
plt.figure(figsize=(15, 8))
sns.scatterplot(x='Tempo', y='Loudness (db)', data=df, color='darkblue')
plt.title('Tempo vs Loudness')
plt.xlabel('Tempo')
plt.ylabel('Loudness (db)')
plt.show()
```

```python
# Boxplot of Song Key vs Popularity
plt.figure(figsize=(18, 8))
sns.boxplot(x='Key', y='Popularity', data=df, palette='Set2')
plt.title('Song Key vs Popularity')
plt.xlabel('Key')
plt.ylabel('Popularity')
plt.show()
```

```python
# Heatmap of Correlation Matrix:
plt.figure(figsize=(20, 8))
# Convert columns to numeric, errors='coerce' will replace non-numeric values with NaN
numeric_df = df.apply(pd.to_numeric, errors='coerce')
# Drop columns with all NaN values after conversion
numeric_df = numeric_df.dropna(axis=1, how='all')
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
