# Univariate Analysis

1.  Univariate analysis is a statistical technique used to analyze the distribution and patterns of a single variable at a time. It's a great starting point to understand individual features in your dataset before moving on to more complex relationships. Here's how we can perform univariate analysis for different data types in your dataset:

2.   Histograms: Visualize the distribution of a single numerical variable by dividing it into intervals (bins) and counting the frequency of data points in each bin. This helps you understand the shape of the distribution (e.g., normal, skewed, uniform).
Box plots: Show the summary statistics (minimum, maximum, median, quartiles) of a numerical variable. They are useful to detect outliers and understand the spread and central tendency of the data.

3.   Descriptive statistics: Calculate measures such as mean, median, standard deviation, and quartiles to gain insights into the central tendency, dispersion, and shape of the distribution.




```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Popularity'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Song Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()

# Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Popularity'], color='lightgreen')
plt.title('Box Plot of Song Popularity')
plt.ylabel('Popularity')
plt.show()

# Descriptive statistics
print(df['Popularity'].describe())
```

# Categorical Features

```python
# Frequency table
key_counts = df['Key'].value_counts()
print(key_counts)

# Bar chart
plt.figure(figsize=(10, 6))
key_counts.plot(kind='bar', color='coral')
plt.title('Frequency Distribution of Song Keys')
plt.xlabel('Key')
plt.ylabel('Frequency')
plt.show()

# Pie chart
plt.figure(figsize=(8, 8))
key_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Song Keys')
plt.ylabel('')  # Remove the default ylabel
plt.show()
```

# Bivariate Analysis

1.   Bivariate analysis explores the relationship between two variables in your dataset. It helps you understand how changes in one variable might affect another and identify potential correlations or dependencies between them.



```python
# Analyzing 'Tempo' vs. 'Loudness'

# Scatter plot
plt.figure(figsize=(15, 8))
sns.scatterplot(x='Tempo', y='Loudness (db)', data=df, color='darkblue')
plt.title('Tempo vs Loudness')
plt.xlabel('Tempo')
plt.ylabel('Loudness (db)')
plt.show()

# Correlation coefficient
correlation = df['Tempo'].corr(df['Loudness (db)'])
print(f"Correlation between Tempo and Loudness: {correlation}")
```

```python
# Analyzing 'Key' vs. 'Popularity'

# Box plot
plt.figure(figsize=(18, 8))
sns.boxplot(x='Key', y='Popularity', data=df, palette='Set2')
plt.title('Song Key vs Popularity')
plt.xlabel('Key')
plt.ylabel('Popularity')
plt.show()
```

```python
# Heatmap of Correlation Matrix:
plt.figure(figsize=(20, 8))
# Convert columns to numeric, errors='coerce' will replace non-numeric values with NaN
numeric_df = df.apply(pd.to_numeric, errors='coerce')
# Drop columns with all NaN values after conversion
numeric_df = numeric_df.dropna(axis=1, how='all')
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
