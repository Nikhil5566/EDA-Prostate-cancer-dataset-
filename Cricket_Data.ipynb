```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('cricket_data_2025.csv')
df.head(11)
```

```python
# Filling the missing data with mode values:

df['Year'] = df['Year'].fillna(df['Year'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(y=df['Player_Name'], order=df['Player_Name'].value_counts().index, palette='coolwarm')
plt.title('Most Common Violation Types')
plt.xlabel('Count')
plt.ylabel('Violation Type')
plt.show()
```

```python
plt.figure(figsize=(17, 7))
sns.histplot(df['Sixes'], bins=10, kde=True, color='blue')
plt.title('Total number of sixes')
plt.xlabel('Stats')
plt.ylabel('Runes')
plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.boxplot( x="Best_Bowling_Match", y='Wickets_Taken', data=df, )
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
