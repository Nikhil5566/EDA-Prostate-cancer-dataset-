# Visualization

```python
plt.figure(figsize = (10, 6))
sns.histplot(df['color_category'])
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.boxplot(df['ripeness'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
