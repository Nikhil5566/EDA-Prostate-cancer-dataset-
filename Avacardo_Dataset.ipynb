# Univariate Analysis

```python
plt.figure(figsize = (10, 6))
sns.histplot(df['saturation']) # Replace 'size' with the actual column name
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.histplot(df['color_category'])
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.boxplot(df['ripeness'])
plt.show()
```

# Bivariate Analysis

```python
# Example 1: Scatter plot between two continuous variables
# Replace 'firmness' and 'saturation' with actual continuous column names from your DataFrame
plt.figure(figsize=(10, 6))
sns.scatterplot(x='firmness', y='saturation', data=df)
plt.title('Scatter Plot of Firmness vs. Saturation')
plt.xlabel('Firmness')
plt.ylabel('Saturation')
plt.show()
```

```python
# Example 2: Box plot between a categorical variable and a continuous variable
# Replace 'color_category' and 'ripeness' with actual column names
plt.figure(figsize=(12, 7))
sns.boxplot(x='color_category', y='ripeness', data=df)
plt.title('Ripeness by Color Category')
plt.xlabel('Color Category')
plt.ylabel('Ripeness')
plt.xticks(rotation=45) # Rotate x-axis labels if they overlap
plt.show()
```

```python
# Example 3: Pair plot for a subset of numerical variables
# Replace these column names with actual numerical column names you want to include
numerical_cols = ['firmness', 'hue', 'saturation', 'ripeness'] # Example numerical columns
sns.pairplot(df[numerical_cols])
plt.suptitle('Pair Plot of Numerical Variables', y=1.02) # Add title above the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
