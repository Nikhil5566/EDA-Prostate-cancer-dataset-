# Visulization

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
sns.countplot(x=df['Species Observed'])
plt.title('Bleaching Severity Distribution', fontsize=15)
plt.xlabel('Bleaching Severity', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

```python
# 1. Correlation Heatmap for numeric columns (if four or more exist)
numeric_df = df.select_dtypes(include=[np.number])
if numeric_df.shape[1] >= 4:
    plt.figure(figsize=(10, 8))
    corr_matrix = numeric_df.corr()
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap of Numeric Features')
    plt.tight_layout()
    plt.show()
```

```python
# 2. Pair Plot
sns.pairplot(numeric_df)
plt.show()

# 3. Histograms for numeric features
numeric_columns = numeric_df.columns
for col in numeric_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(numeric_df[col], bins=30, kde=True)
    plt.title(f'Histogram of {col}')
    plt.tight_layout()
    plt.show()
```

```python
# 4. Pie Chart / Count plot for Bleaching Severity
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Bleaching Severity', palette='Set2')
plt.title('Bleaching Severity Counts')
plt.tight_layout()
plt.show()

# 5. Count plot for Marine Heatwave (boolean)
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Marine Heatwave', palette='Set1')
plt.title('Marine Heatwave Occurrence')
plt.tight_layout()
plt.show()

# 6. Box Plot for SST and pH Level across Bleaching Severity categories
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Bleaching Severity', y='SST (°C)', palette='cool')
plt.title('SST Distribution by Bleaching Severity')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Bleaching Severity', y='pH Level', palette='cool')
plt.title('pH Level Distribution by Bleaching Severity')
plt.tight_layout()
plt.show()
```
