# Handling Outliers

```python
# 1. Identify Outliers

for column in df.select_dtypes(include=np.number).columns:
  plt.figure(figsize=(8,6))
  sns.boxplot(x=df[column])
  plt.title(f'Box plot of{column}')
  plt.show()
```

```python
# Using IQR (Interquartile Range):

# Select only numeric columns for IQR calculation
numeric_df = df.select_dtypes(include=np.number)

Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers in numeric columns
outliers = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)

# If you need to see which rows in the original DataFrame are outliers:
outlier_rows = df[outliers.reindex(df.index, fill_value=False)]
```

```python
# Handle Outliners

df_no_outliers = df[~outliers]
```

# Univariate analysis:

```python
# 1. Numerical Variables:

for column in df.select_dtypes(include=np.number).columns:
    plt.figure(figsize=(8, 6))
    plt.hist(df[column], bins=10)  # Adjust bins as needed
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Categorial Variable

for column in df.select_dtypes(include=['object']).columns:
  plt.figure(figsize=(15, 8))
  sns.countplot(x=df[column])
  plt.title(f'Bar Chart of {column}')
  plt.xlabel(column)
  plt.ylabel('Count')
  plt.xticks(rotation=90, ha='right')
  plt.show()
```

# Feature Engineering

```python
# 5. Feature Engineering

# Interaction Term
df['legs_x_aquatic'] = df['legs'] * df['aquatic']

# Polynomial Feature
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, include_bias=False)
hair_poly = poly.fit_transform(df[['hair']])
df['hair_squared'] = hair_poly[:, 1]

# Domain-Specific Feature
# Before feature scaling, create the 'bird_like' feature
df['bird_like'] = df['feathers'].astype(int) & df['eggs'].astype(int)

# Feature Encoding (If needed - Example with one-hot encoding)
# Check if 'type' column exists before applying get_dummies
if 'type' in df.columns:
    df = pd.get_dummies(df, columns=['type'], prefix=['type'])
else:
    print("Column 'type' not found in DataFrame. Skipping one-hot encoding.")


# Feature Scaling (If needed - Example with StandardScaler)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numerical_features = ['hair', 'feathers', 'eggs', 'milk', 'airborne', 'aquatic', 'predator', 'toothed', 'backbone', 'breathes', 'venomous', 'fins', 'legs', 'tail', 'domestic', 'catsize', 'legs_x_aquatic', 'hair_squared']  # Include new features
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```
