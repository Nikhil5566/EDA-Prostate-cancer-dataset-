```python
df['Year'] = df['Year'].fillna(df['Year'].max())
```

```python
# Here the publisher is an OBJECT value but it can not be converted into any other astype because it had an nintado error that means it was an non numerical columns so we have used category instead:

df['Publisher'] = df['Publisher'].astype('category')
```

```python
df['Publisher'] = df['Publisher'].fillna(df['Publisher'].mode()[0])
```

# Visualization

```python
# 1. Sales by Region (North America, Europe, Japan, Other)
region_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
plt.figure(figsize=(8, 6))
sns.barplot(x=region_sales.index, y=region_sales.values, palette='muted')
plt.title('Total Sales by Region')
plt.ylabel('Total Sales (in millions)')
plt.xlabel('Region')
plt.show()
```

```python
# Set Seaborn style for better aesthetics
sns.set(style="whitegrid")

# 1. Distribution of Global Sales
plt.figure(figsize=(10, 6))
sns.histplot(df['Global_Sales'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Global Sales')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Sales Trends Over Years (Global Sales)
plt.figure(figsize=(12, 6))
yearly_sales = df.groupby('Year')['Global_Sales'].sum()
sns.lineplot(x=yearly_sales.index, y=yearly_sales.values, marker='o', color='orange')
plt.title('Global Sales Trends Over Years')
plt.xlabel('Year')
plt.ylabel('Total Global Sales (in millions)')
plt.show()
```

```python
sales_columns = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
corr_matrix = df[sales_columns].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt=".2f")
plt.title('Correlation Heatmap of Sales by Region')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
