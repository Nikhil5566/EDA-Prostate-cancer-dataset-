# Visualisation

```python
sns.pairplot(df)
```

```python
sns.histplot(df['Tuition_USD'], bins=20, kde=True)
plt.title('Distribution of Tuition Fees (USD)')
plt.xlabel('Tuition (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(df['Living_Cost_Index'], bins=20, kde=True)
plt.title('Distribution of Living Cost Index')
plt.xlabel('Living Cost Index')
plt.ylabel('Frequency')
plt.show()
```

```python
avg_tuition_country = df.groupby('Country')['Tuition_USD'].mean().sort_values()
avg_tuition_country.plot(kind='barh', figsize=(10,8))
plt.title('Average Tuition Fee by Country')
plt.xlabel('Average Tuition (USD)')
plt.ylabel('Country')
plt.show()
```

```python
sns.scatterplot(x='Living_Cost_Index', y='Tuition_USD', data=df)
plt.title('Tuition Fees vs. Living Cost Index')
plt.xlabel('Living Cost Index')
plt.ylabel('Tuition (USD)')
plt.show()
```

```python
program_counts = df['Program'].value_counts()
program_counts.plot(kind='bar', figsize=(10,6))
plt.title('Number of Universities by Program')
plt.xlabel('Program')
plt.ylabel('Number of Universities')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
