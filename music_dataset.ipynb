# Visualization

```python
plt.figure(figsize=(15, 8))
sns.boxplot(x=df["Artist"], y=df["Streams"], palette="coolwarm")
plt.title("Deviced used as per the location", fontsize=14)
plt.xlabel("Devices")
plt.ylabel("Location")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
