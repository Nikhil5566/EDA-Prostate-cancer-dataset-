# Visualization

```python
year_columns = df.columns[2:]

# Loop over each year column to plot the top 10 ranking for that year
for year in year_columns:
    # Create a temporary DataFrame with country names and the current year's values
    df_year = df[['Country Name', year]].copy()
    df_year.columns = ['Country Name', 'Value']

    # Convert the 'Value' column to numeric; non-numeric entries become NaN
    df_year['Value'] = pd.to_numeric(df_year['Value'], errors='coerce')

    # Drop rows where the value is missing
    df_year = df_year.dropna(subset=['Value'])

    # Sort by the value in descending order and take the top 10
    df_top10 = df_year.sort_values(by='Value', ascending=False).head(10)

    # Create a 'Rank' column (1 for highest, 10 for lowest)
    df_top10['Rank'] = range(1, len(df_top10) + 1)

    # Plot a horizontal bar chart
    plt.figure(figsize=(8, 6))
    plt.barh(df_top10['Country Name'], df_top10['Value'], color='skyblue')
    plt.xlabel('Value')
    plt.title(f'Top 10 Countries in {year}')

    # Annotate each bar with its rank
    for i, (val, rank) in enumerate(zip(df_top10['Value'], df_top10['Rank'])):
        plt.text(val, i, f'{val}', va='center', ha='left')

    # Invert y-axis so the highest value (rank 1) is at the top
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
```
