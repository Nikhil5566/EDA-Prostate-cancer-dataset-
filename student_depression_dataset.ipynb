```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette('rocket')
import warnings
warnings.filterwarnings('ignore')
```

```python
filtered_age = df[df['Age'] > 20]
filtered_age.head(11)
```

```python
filtered_Study_Satisfaction	= df[df['Study Satisfaction'] > 2]
filtered_Study_Satisfaction.head(11)
```

```python
df.groupby('Gender').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Create a histogram
plt.subplots(figsize=(15, 7))
plt.hist(filtered_age['Age'], bins=30, color='blue', edgecolor='black')

# Labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Age")

# Show the plot
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Financial Stress'].value_counts()
    for x_label, grp in df.groupby('Have you ever had suicidal thoughts ?')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Have you ever had suicidal thoughts ?')
_ = plt.ylabel('Financial Stress')
```

```python
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Sleep Duration'].value_counts()
    for x_label, grp in df.groupby('Gender')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Gender')
_ = plt.ylabel('Sleep Duration')
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
