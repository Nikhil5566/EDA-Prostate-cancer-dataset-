```python
# Handling missing values
df['Type 2'] = df['Type 2'].fillna(df['Type 2'].mode()[0])
df['gender_male_ratio'] = df['gender_male_ratio'].fillna(df['gender_male_ratio'].min())
df = df.drop(columns = ['Forms'])
```

```python
# Pair Plot for select combat stats
selected_stats = ['HP', 'Attack', 'Defense', 'Speed', 'Base_Stats']
sns.pairplot(df[selected_stats])
plt.suptitle("Pair Plot for Select Stats", y=1.02)
plt.show()
```

```python
# 2. Histogram of Age and Income
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
sns.histplot(df['Attack'], kde=True, color='teal')
plt.title('Attack Distribution')

plt.subplot(1, 2, 2)
sns.histplot(df['Defense'], kde=True, color='coral')
plt.title('Defense Distribution')
plt.tight_layout()
plt.show()
```

```python
# 3. Box Plot for Income Distribution
plt.figure(figsize=(8,6))
sns.boxplot(y='Speed', data=df, color='lightblue')
plt.title('Box Plot of Income')
plt.tight_layout()
plt.show()
```

```python
# 4. Violin Plot for Age distribution across Sex
plt.figure(figsize=(8,6))
sns.violinplot(x='fire_weakness', y='water_weakness', data=df, palette='Set2')
plt.title('Violin Plot: Weakness')
plt.tight_layout()
plt.show()
```

```python
# 5. Strip Plot: Income across Education levels
plt.figure(figsize=(15,8))
sns.stripplot(x='Classification_info', y='number_not_effective', data=df, jitter=True, palette='Spectral')
plt.title('Strip Plot: Income Distribution across Education Levels')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
