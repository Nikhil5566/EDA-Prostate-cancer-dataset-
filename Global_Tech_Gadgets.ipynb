```python
# Smartphone Sales (Millions)

from matplotlib import pyplot as plt
df['Smartphone Sales (Millions)'].plot(kind='hist', bins=20, title='Smartphone Sales (Millions)')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Smartphone Sales (Millions) vs Laptop Shipments (Millions)

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Smartphone Sales (Millions)', y='Laptop Shipments (Millions)', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Extract numeric columns for correlation analysis
numeric_df = df.select_dtypes(include=[np.number])
print('Numeric columns:', list(numeric_df.columns))

# Create a correlation heatmap if we have >= 4 numeric columns
if numeric_df.shape[1] >= 4:
    plt.figure(figsize=(10, 6))
    corr = numeric_df.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap of Numeric Features')
    plt.show()
else:
    print('Not enough numeric columns for a correlation heatmap.')
```

```python
# Histograms for each numeric column
numeric_df.hist(bins=15, figsize=(15, 10), layout=(3, 3));
plt.suptitle('Histograms of Numeric Features', y=1.02)
plt.show()
```

```python
# Pair plot to visualize pairwise relationships between numeric variables
sns.pairplot(numeric_df)
plt.suptitle('Pair Plot of Numeric Features', y=1.02)
plt.show()
```

```python
# Count plot for the categorical variable 'Country' (Pie chart alternative could be used, but countplot gives clear counts)
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Country', order=df['Country'].value_counts().index)
plt.title('Count Plot of Countries')
plt.xticks(rotation=45)
plt.show()
```
