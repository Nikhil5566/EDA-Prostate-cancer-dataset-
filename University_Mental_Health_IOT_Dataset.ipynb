# Handling Outliers

```python
Q1 = df['sleep_hours'].quantile(0.25)
Q3 = df['sleep_hours'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 - 1.5 * IQR

# Outliers Before:
outliers_before = df[(df['sleep_hours'] < lower_bound) | (df['sleep_hours'] > upper_bound)].shape[0]
print(f"Number of outliers in '{'sleep_hours'}' before handling: {outliers_before}")

# Handle outliers by capping
df['sleep_hours'] = df['sleep_hours'].clip(lower=lower_bound, upper=upper_bound)

# Count outliers after handling (should be 0 if capped)
outliers_after = df[(df['sleep_hours'] < lower_bound) | (df['sleep_hours'] > upper_bound)].shape[0]
print(f"Number of outliers in '{'sleep_hours'}' after handling: {outliers_after}")
```

# Visualization

```python
# 1. Histogram/Distribution Plot
plt.figure(figsize=(15, 8))
sns.histplot(data=df, x='sleep_hours', kde=True, bins=20)
plt.title('Distribution of Sleep Hours')
plt.xlabel('Sleep Hours')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Countplot
plt.figure(figsize=(15, 8))
sns.countplot(data=df, x='crowd_density')
plt.title('Count of Students by crowd_density')
plt.xlabel('crowd_density')
plt.ylabel('Count')
plt.show()
```

```python
# 3. Box Plot (comparing across categories)
plt.figure(figsize=(15, 8))
sns.boxplot(data=df, x='lighting_lux', y='noise_level_db')
plt.title('Boxplot for lighting and noise')
plt.xlabel('lighting_lux')
plt.ylabel('noise_level_db')
plt.show()
```

```python
# 4. Scatter Plot
plt.figure(figsize=(15, 8))
sns.scatterplot(data=df, x='mood_score', y='mental_health_status')
plt.title(' Mood Score vs. Mental Health')
plt.xlabel('Sleep Hours')
plt.ylabel('Stress Level')
plt.show()
```

```python
# 5. Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='location_id', y='temperature_celsius')
plt.title('Location id by Temperature Celsius')
plt.xlabel('Location ID')
plt.ylabel('Temperature Celsius')
plt.show()
```

```python
# Heatmap:
plt.figure(figsize = (15, 8))
numerical_df = df.select_dtypes(include=np.number)
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
