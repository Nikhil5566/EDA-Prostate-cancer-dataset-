# Outliers

```python
# Identify numerical features
numerical_features = ['price', 'year', 'odometer']

# Calculate IQR and bounds
for feature in numerical_features:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    print(f"Feature: {feature}, Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")

# Identify outliers
outliers = df[(df[numerical_features] < lower_bound) | (df[numerical_features] > upper_bound)].any(axis=1)

# Handle outliers (choose one option)
# 1. Removal:
df_no_outliers = df[~outliers]
```

```python
# Numerical variables
numerical_features = ['price', 'year', 'odometer']

# Descriptive statistics for numerical features
for feature in numerical_features:
    print(f"Descriptive statistics for {feature}:\n{df[feature].describe()}\n")

# Histograms for numerical features
for feature in numerical_features:
    plt.hist(df[feature], bins=20)
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {feature}')
    plt.show()

# Box plots for numerical features
for feature in numerical_features:
    plt.boxplot(df[feature])
    plt.ylabel(feature)
    plt.title(f'Box Plot of {feature}')
    plt.show()

# Categorical variables
categorical_features = ['manufacturer', 'model', 'fuel', 'transmission']

# Frequency tables for categorical features
for feature in categorical_features:
    print(f"Frequency table for {feature}:\n{df[feature].value_counts()}\n")

# Bar charts for categorical features
for feature in categorical_features:
    sns.countplot(x=feature, data=df)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(f'Distribution of {feature}')
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.show()
```

```python
# Numerical vs. Numerical
plt.scatter(df['price'], df['odometer'])
plt.xlabel('Price')
plt.ylabel('Odometer')
plt.title('Scatter Plot: Price vs. Odometer')
plt.show()

print(df[['price', 'year', 'odometer']].corr())

# Categorical vs. Categorical
print(pd.crosstab(df['manufacturer'], df['fuel']))

sns.countplot(x='manufacturer', hue='fuel', data=df)
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.title('Stacked Bar Chart: Manufacturer vs. Fuel')
plt.xticks(rotation=90)
plt.show()

# Numerical vs. Categorical
sns.boxplot(x='fuel', y='price', data=df)
plt.xlabel('Fuel')
plt.ylabel('Price')
plt.title('Box Plot: Price vs. Fuel')
plt.show()

sns.violinplot(x='fuel', y='price', data=df)
plt.xlabel('Fuel')
plt.ylabel('Price')
plt.title('Violin Plot: Price vs. Fuel')
plt.show()
```

# Visualization

```python
# 1. Scatter plot for Price vs. Odometer
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
sns.scatterplot(x='odometer', y='price', data=df)
plt.title('Scatter Plot: Price vs. Odometer')
plt.xlabel('Odometer')
plt.ylabel('Price')
plt.show()

# 2. Heatmap for Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df[['price', 'year', 'odometer']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()

# 3. Stacked Bar Chart for Manufacturer vs. Fuel
plt.figure(figsize=(10, 6))
sns.countplot(x='manufacturer', hue='fuel', data=df)
plt.title('Stacked Bar Chart: Manufacturer vs. Fuel')
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Fuel Type')
plt.show()

# 4. Box Plot for Price vs. Fuel
plt.figure(figsize=(8, 6))
sns.boxplot(x='fuel', y='price', data=df)
plt.title('Box Plot: Price vs. Fuel')
plt.xlabel('Fuel')
plt.ylabel('Price')
plt.show()

# 5. Violin Plot for Price vs. Fuel
plt.figure(figsize=(8, 6))
sns.violinplot(x='fuel', y='price', data=df)
plt.title('Violin Plot: Price vs. Fuel')
plt.xlabel('Fuel')
plt.ylabel('Price')
plt.show()
```
