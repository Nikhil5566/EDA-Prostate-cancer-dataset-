```python
# Filling director, cast, country with mode

df.drop(['director', 'cast', 'country' ], axis=1, inplace=True)
```

# Visualization

```python
# Plot a count of content types (Movie vs TV Show)
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='type', palette='viridis')
plt.title('Count of Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Histogram for release_year
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='release_year', bins=30, kde=True, color='blue')
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=df, y='rating', order=df['rating'].value_counts().index, palette='mako')
plt.title('Distribution of Ratings')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```
