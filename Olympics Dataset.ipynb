```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Olympics (1896-2024).csv')
df.head(11)
```

```python
df['Year'].plot(kind='hist', bins=20, title='Year')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
plt.figure(figsize=(20, 8))
sns.countplot(x=df['Total'], palette="viridis")
plt.title("Overall", fontsize=14)
plt.xlabel("Years")
plt.ylabel("Medals")

plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.countplot(x=df['Rank'], palette="viridis")
plt.title("Ranks of the countries", fontsize=14)
plt.xlabel("X-Axis")
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(x=df["Year"], y=df["Gold"], palette="coolwarm")
plt.title("Years vs Gold", fontsize=14)
plt.xlabel("Years")
plt.ylabel("Golds")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(x=df["Year"], y=df["Silver"], palette="coolwarm")
plt.title("Years vs Silver", fontsize=14)
plt.xlabel("Years")
plt.ylabel("Silvers")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(x=df["Year"], y=df["Bronze"], palette="coolwarm")
plt.title("Years vs Bronze", fontsize=14)
plt.xlabel("Years")
plt.ylabel("Bronze")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.boxplot(x=df["Year"], y=df["Total"], palette="coolwarm")
plt.title("Total years vs Gold", fontsize=14)
plt.xlabel("Years")
plt.ylabel("Total")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
