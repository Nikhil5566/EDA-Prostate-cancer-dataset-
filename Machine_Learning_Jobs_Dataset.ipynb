```python
df['company_address_region'] = df['company_address_region'].fillna(df['company_address_region'].mode()[0])
df = df.dropna()
```

# Visualisation

```python
# Create a histogram of the salary feature
plt.figure(figsize=(10, 6))  # Adjust figure size if needed
sns.histplot(df['job_posted_date'], bins=20, kde=True)  # kde adds a kernel density estimate
plt.title('Publised Jobs')
plt.xlabel('Year')
plt.ylabel('Applied Frequency')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x='company_address_locality', y='company_address_region', data=df)
plt.title('Salary vs. Experience')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(15, 8))  # Adjust figure size as needed
sns.boxplot(x='company_description', y='company_website', data=df)
plt.title('Salary Distribution by Job Title')
plt.xlabel('Job Title')
plt.ylabel('Salary')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels if needed
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
