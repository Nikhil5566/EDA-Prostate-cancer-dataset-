# Visualization

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Mental Stress Level'], bins=10, kde=True, color='blue')
plt.title('Distribution of Mental Stress Levels')
plt.xlabel('Stress Level')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Mental Stress Level'], y=df['Study Hours Per Week'], palette='coolwarm')
plt.title('Study Hours vs. Mental Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Study Hours Per Week')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x=df['Mental Stress Level'], y=df['Sleep Duration (Hours per night)'], palette='pastel')
plt.title('Sleep Duration vs. Mental Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Sleep Duration (Hours)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=df['Mental Stress Level'], y=df['Social Media Usage (Hours per day)'], palette='coolwarm')
plt.title('Social Media Usage vs. Mental Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Social Media Usage (Hours per day)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=df['Family Support  '], y=df['Mental Stress Level'], palette='Greens')
plt.title('Family Support vs. Mental Stress Level')
plt.xlabel('Family Support Level')
plt.ylabel('Average Stress Level')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
