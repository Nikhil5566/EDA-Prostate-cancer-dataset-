```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('IMDb 2024 Movies TV Shows.csv')
df.head(10)
```

```python
missing_values = df.isnull().sum()

duplicated_values = df.duplicated().sum()

print("Missing Values:")
print(missing_values)
print("\nDuplicated Values:", duplicated_values)
```

```python
plt.figure(figsize = (12, 8 ))
scatter = plt.scatter(df['Movie_Name'],
                      df['Budget'],
                      alpha = 0.6,
                      cmap = 'viridus')
plt.legend(*scatter.legend_elements(),
           loc = 'upper left',
           title = 'Budget')

plt.xticks(rotation = 45)
plt.title('Scatter-plot-Movie', fontsize = 14, weight = 'bold')
plt.xlabel('Movie-Name', weight = 'bold')
plt.ylabel('Budget')

plt.show()
```
