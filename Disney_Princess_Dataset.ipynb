# Visualization

```python
df['IsIconic'].value_counts()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12,6))
sns.countplot(x=df['IsIconic'], palette='winter')
plt.title('Dsitribution of Iconic', fontsize=16)
plt.xlabel('Iconic', fontsize=13)
plt.ylabel('Count', fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x=df['PrincessName'], palette='jet_r')
plt.title('Distribution of PrincessName', fontsize=16)
plt.xlabel('PrincessName', fontsize=13)
plt.ylabel('Popularity Score', fontsize=13)
plt.xticks(rotation=50)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x=df['PrincessName'], palette='mako', hue=df['HasSoloSong'])
plt.title('Princess Name By HasSolo Song', fontsize=16)
plt.xlabel('PrincessName', fontsize=13)
plt.ylabel('Count', fontsize=13)
plt.xticks(rotation=50)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x=df['PrincessName'], palette='jet_r', hue=df['IsIconic'])
plt.title('Princess Name By IsIconic', fontsize=16)
plt.xlabel('PrincessName', fontsize=13)
plt.ylabel('Count', fontsize=13)
plt.xticks(rotation=50)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x=df['OutfitPrimaryColor'], y=df['PopularityScore'])
plt.title('Outfit Primary Color Vs Popularity Score', fontsize=16)
plt.xlabel('OutfitPrimaryColor', fontsize=13)
plt.ylabel('PopularityScore', fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x=df['NumberOfSongs'], y=df['BoxOfficeMillions'], palette='Blues')
plt.title('Number Of Songs Vs BoxOffice Millions', fontsize=16)
plt.xlabel('NumberOfSongs', fontsize=13)
plt.ylabel('BoxOfficeMillions', fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='HasSoloSong', y='PopularityScore', data=df)
plt.title('Solo Song Vs Popularity Score', fontsize=15)
plt.xlabel('HasSoloSong', fontsize=13)
plt.ylabel('PopularityScore', fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
