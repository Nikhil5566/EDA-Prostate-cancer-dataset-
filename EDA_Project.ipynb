# Percentage of missing values per columns:

```python
df.isna().sum().sort_values(ascending=False) # This is used for arranging the values per missing values as for which have more error or null values
```

```python
# missing_percentage[missing_percentage != 0] # To check how many columns have missing values
```

```python
# missing percentage:

missing_percentage = df.isna().sum().sort_values(ascending=True)/len(df)
missing_percentage
```

```python
# missing percentage:

missing_percentage = df.isna().sum().sort_values(ascending=True)/len(df)

# Check if there are any missing values before plotting
if (missing_percentage != 0).any():
    # ploting as a graph:
    missing_percentage[missing_percentage != 0].plot(kind='barh')
else:
    print("No missing values to plot.")
```
