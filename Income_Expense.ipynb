# Visualization

```python
# Count plot for Geographic distribution:

plt.figure(figsize=(14, 6))
sns.countplot(data=df, x='GEO', order=df['GEO'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Distribution of Records by Geographic Location')
plt.tight_layout()
plt.show()
```

```python
# Histogram for Income and Expense
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
sns.histplot(df['Income'], kde=True, ax=ax[0], color='skyblue')
ax[0].set_title('Income Distribution')

sns.histplot(df['Expense'], kde=True, ax=ax[1], color='salmon')
ax[1].set_title('Expense Distribution')
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=df, y='Statistic', order=df['Statistic'].value_counts().index, palette='mako')
plt.title('Distribution of Ratings')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
