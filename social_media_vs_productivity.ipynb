```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df['daily_social_media_time'] = df['daily_social_media_time'].fillna(df['daily_social_media_time'].mode()[0])
df['perceived_productivity_score'] = df['perceived_productivity_score'].fillna(df['perceived_productivity_score'].mode()[0])
df['actual_productivity_score'] = df['actual_productivity_score'].fillna(df['actual_productivity_score'].mode()[0])
df['stress_level'] = df['stress_level'].fillna(df['stress_level'].mode()[0])
df['sleep_hours'] = df['sleep_hours'].fillna(df['sleep_hours'].mode()[0])
df['screen_time_before_sleep'] = df['screen_time_before_sleep'].fillna(df['screen_time_before_sleep'].mode()[0])
df['job_satisfaction_score'] = df['job_satisfaction_score'].fillna(df['job_satisfaction_score'].mode()[0])
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# Example using a box plot for 'daily_social_media_time'
sns.boxplot(x=df['daily_social_media_time'])
plt.title('Box plot of Daily Social Media Time')
plt.show()

# Example using a scatter plot between 'daily_social_media_time' and 'actual_productivity_score'
plt.scatter(df['daily_social_media_time'], df['actual_productivity_score'])
plt.xlabel('Daily Social Media Time')
plt.ylabel('Actual Productivity Score')
plt.title('Scatter plot of Social Media Time vs. Productivity')
plt.show()
```

```python
# 2. Z-Score:
from scipy import stats

z_scores = np.abs(stats.zscore(df['daily_social_media_time']))
outlier_indices = np.where(z_scores > 3)
print("Indices of outliers in 'daily_social_media_time' based on Z-score:")
print(outlier_indices)
```

```python
Q1 = df['daily_social_media_time'].quantile(0.25)
Q3 = df['daily_social_media_time'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['daily_social_media_time'] < lower_bound) | (df['daily_social_media_time'] > upper_bound)]
print("Outliers in 'daily_social_media_time' based on IQR:")
print(outliers)
```

```python
df_cleaned = df[(df['daily_social_media_time'] >= lower_bound) & (df['daily_social_media_time'] <= upper_bound)]
print("Shape of DataFrame after removing outliers:")
print(df_cleaned.shape)
```

# Univariate Analysis

```python
# Measures of Dispersion (Spread):

variance_value = df['daily_social_media_time'].var()
std_dev_value = df['daily_social_media_time'].std()
range_value = df['daily_social_media_time'].max() - df['daily_social_media_time'].min()
Q1 = df['daily_social_media_time'].quantile(0.25)
Q3 = df['daily_social_media_time'].quantile(0.75)
iqr_value = Q3 - Q1

print(f"Variance: {variance_value}")
print(f"Standard Deviation: {std_dev_value}")
print(f"Range: {range_value}")
print(f"IQR: {iqr_value}")
```

```python
# Frequency Distribution:

plt.hist(df['daily_social_media_time'], bins=20) # You can adjust the number of bins
plt.xlabel('Daily Social Media Time')
plt.ylabel('Frequency')
plt.title('Histogram of Daily Social Media Time')
plt.show()
```

# Bivariate Analysis

```python
# Scatter Plot:

plt.scatter(df['daily_social_media_time'], df['actual_productivity_score'])
plt.xlabel('Daily Social Media Time')
plt.ylabel('Actual Productivity Score')
plt.title('Scatter plot of Social Media Time vs. Actual Productivity Score')
plt.show()
```

```python
# Correlation Coefficient

correlation_pearson = df['daily_social_media_time'].corr(df['actual_productivity_score'], method='pearson')
print(f"Pearson correlation between daily_social_media_time and actual_productivity_score: {correlation_pearson}")
```

```python
# Box-plots:

sns.boxplot(x='uses_focus_apps', y='daily_social_media_time', data=df)
plt.title('Daily Social Media Time by Category')
plt.show()
```

```python
# Grouped Bar Charts:

grouped_means = df.groupby('uses_focus_apps')['daily_social_media_time'].mean()
print("Mean daily_social_media_time by category:")
print(grouped_means)

# To visualize as a bar chart
grouped_means.plot(kind='bar')
plt.xlabel('uses_focus_apps')
plt.ylabel('Mean Daily Social Media Time')
plt.title('Mean Daily Social Media Time by Category')
plt.show()
```

# Visualization

```python
# Histogram:

plt.figure(figsize = (15, 8))
plt.hist(df['daily_social_media_time'], bins=20, edgecolor='black') # Adding edgecolor for better visual separation
plt.xlabel('Daily Social Media Time (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Daily Social Media Time')
plt.show()
```

```python
# Scatter Plot:

plt.figure(figsize=(15, 8))
plt.scatter(df['daily_social_media_time'], df['perceived_productivity_score'], alpha=0.5) # Adding alpha for transparency
plt.xlabel('Daily Social Media Time (minutes)')
plt.ylabel('Perceived Productivity Score')
plt.title('Scatter Plot of Daily Social Media Time vs. Perceived Productivity Score')
plt.show()
```

```python
# Count-Plot:

plt.figure(figsize=(15, 8))
sns.countplot(x='uses_focus_apps', data=df)
plt.xlabel('Uses Focus Apps')
plt.ylabel('Count')
plt.title('Count of Users Using Focus Apps')
plt.show()
```

```python
# Heatmap of Correlations:

 # Select only numerical columns for correlation
numerical_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```
