```python
# Stacked Area Chart

labels = ['Country','Fossil Fuel Dependency']   # Its should follow the order that it is writtin as stackplot df :
sns.set_style ('white')              # Changing the background color:
plt.figure(figsize = (12, 6))
plt.stackplot(df['Country'],
              df['Fossil Fuel Dependency (%)'],
              # color = colors,
              edgecolor = 'none')


plt.xticks(df['Country'], rotation= 45)
plt.legend(labels = labels, loc = 'upper right')
plt.ylabel('Fossil Fuel', fontsize = 13)
plt.title('', fontsize = 14, weight ='bold')
sns.despine()    # Removing  the top and right ouline in the figure:
#plt.savefig('Stackplot')
plt.show()
```

```python
# Year vs Total Energy Consumption (TWh)

def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Year']
  ys = series['Total Energy Consumption (TWh)']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Year', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Year')
_ = plt.ylabel('Total Energy Consumption (TWh)')
```

```python
# Total Energy Consumption (TWh) vs Per Capita Energy Use (kWh)

df.plot(kind='scatter', x='Total Energy Consumption (TWh)', y='Per Capita Energy Use (kWh)', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Boxploting

plt.figure(figsize=(10, 6))
sns.boxplot(x=df["Industrial Energy Use (%)"], y=df["Household Energy Use (%)"], palette="coolwarm")
plt.title("Energy Usage", fontsize=14)
plt.xlabel("Industrial Energy Usage")
plt.ylabel("Household Energy Usage")
plt.show()
```

```python
sns.boxplot( x="Country", y='Fossil Fuel Dependency (%)', data=df, )
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
