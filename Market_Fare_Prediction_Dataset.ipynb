```python
# Example: Assuming you want to plot a subset of data where 'Market_share' is between 0 and 10
subset_df = df[(df['Market_share'] >= 0) & (df['Market_share'] <= 50)].copy()

# Now plot using the subset_df
plt.figure(figsize=(10, 6))
sns.histplot(subset_df['Market_share'], kde=True)
plt.title('Distribution of Market Share (0-10)')
plt.xlabel('Market Share')
plt.ylabel('Frequency')
plt.show()
```

# Handling Duplicated Values:

```python
# Check the number of duplicates before dropping
print("Number of duplicated rows before dropping:", df.duplicated().sum())

# Drop duplicate rows
df.drop_duplicates(inplace=True)

# Check the number of duplicates after dropping
print("Number of duplicated rows after dropping:", df.duplicated().sum())
```

# Handling Outliers:

```python
Q1 = df['Market_share'].quantile(0.25)
Q3 = df['Market_share'].quantile(0.75)

lower_bound = Q1 - 1.5 * (Q3 - Q1)
upper_bound = Q3 + 1.5 * (Q3 - Q1)

outliers = df[(df['Market_share'] < lower_bound) | (df['Market_share'] > upper_bound)]

print(f"Number of outliers: {len(outliers)}")
```

# Univariate Analysis:

```python
# 1. Summary Statistics
print("Summary Statistics for Market_share:")
print(df['Market_share'].describe())

# 2. Distribution Visualization (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df['Market_share'], kde=True)
plt.title('Distribution of Market Share')
plt.xlabel('Market Share')
plt.ylabel('Frequency')
plt.show()

# 3. Distribution Visualization (Box Plot)
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Market_share'])
plt.title('Box Plot of Market Share')
plt.xlabel('Market Share')
plt.ylabel('Value')
plt.show()
```

# Bivariate Analysis:

```python
# Scatter Plot:

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Average_Fare', y='Market_share', data=df)
plt.title('Scatter Plot of Average Fare vs. Market Share')
plt.xlabel('Average Fare')
plt.ylabel('Market Share')
plt.show()
```

```python
# Box-Ploting:

plt.figure(figsize = (10, 6))
sns.boxplot(x = 'Slot', y = 'Market_share', data=df)
plt.xlabel('Category')
plt.ylabel('Market Share')
plt.xticks(rotation=45) # Rotate labels if they overlap
plt.show()
```

# Visualization:

```python
# Histoplot:

plt.figure(figsize=(10, 6))
sns.histplot(df['Average_Fare'], kde = True)
plt.title('Histogram of Average Fare')
plt.xlabel('Average Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box-Plot for Market Share:

plt.figure(figsize=(10, 6))
sns.boxplot(x = 'Non_Stop', y = 'Carrier_freq', data = df)
plt.xlabel('Market Share')
plt.ylabel('Non_Stop')
plt.xticks(rotation = 45)
plt.show()
```

```python
# Scatter-Plot:

plt.figure(figsize=(10,6))
sns.scatterplot(x='Average_Fare', y='Carrier_freq', data=df)
plt.title('Scatter Plot of Average Fare vs. Market Share')
plt.xlabel('Average Fare')
plt.ylabel('Market Share')
plt.show()
```

```python
# Heatmap:

numerical_cols = df.select_dtypes(include=np.number).columns
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(15, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```
