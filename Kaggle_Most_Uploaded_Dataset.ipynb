# Handling Outliers:

```python
Q1 = df['Upvote'].quantile(0.25)
Q3 = df['Upvote'].quantile(0.25)

IQR = Q3 - Q1

lower_bound = Q1 * 1.5 - IQR
upper_bound = Q3 * 1.5 - IQR

df_outliers = df[(df['Upvote'] >= lower_bound) | (df['Upvote'] <= upper_bound)].copy()
df_outliers

print('Numbers of outliers:', len(df_outliers))
```

# Univariate Analysis

```python
# Univariate Analysis for Continuous Variable 'Upvote'
print("Summary Statistics for Upvote:")
print(df['Upvote'].describe())

plt.figure(figsize=(8, 6))
sns.histplot(df['Upvote'], kde=True, bins=20)
plt.title('Distribution of Upvotes')
plt.xlabel('Upvote')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Upvote'])
plt.title('Box Plot of Upvotes')
plt.ylabel('Upvote')
plt.show()

# Univariate Analysis for Categorical Variable 'Data_Type'
print("\nFrequency Table for Data_Type:")
print(df['Data_Type'].value_counts())

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Data_Type')
plt.title('Count of Datasets by Data Type')
plt.xlabel('Data Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Bivariate analysis

```python
# Bivariate Analysis: Relationship between 'Upvote' and 'Views' (Two Continuous Variables)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='File_Count', y='Upvote')
plt.title('Upvotes vs. File_Count')
plt.xlabel('File_Count')
plt.ylabel('Upvotes')
plt.show()

# Calculate the correlation coefficient
correlation = df['Upvote'].corr(df['File_Count'])
print(f"Correlation between Upvote and File_Count: {correlation}")

# Bivariate Analysis: Relationship between 'Upvote' and 'Data_Type' (One Continuous, One Categorical)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Data_Type', y='Upvote')
plt.title('Upvotes by Data Type')
plt.xlabel('Data Type')
plt.ylabel('Upvotes')
plt.xticks(rotation=45)
plt.show()

# Bivariate Analysis: Relationship between 'Usability' and 'Data_Type' (One Continuous, One Categorical)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Data_Type', y='Usability')
plt.title('Usability by Data Type')
plt.xlabel('Data Type')
plt.ylabel('Usability')
plt.xticks(rotation=45)
plt.show()
```

# Visualization

```python
# Exploratory Data Analysis (EDA)

# List of numeric columns (we use only numeric columns for correlation heatmaps)
numeric_df = df.select_dtypes(include=[np.number])
print('Numeric columns for EDA:')
print(numeric_df.columns.tolist())

# If there are at least 4 numeric columns, produce a correlation heatmap
if numeric_df.shape[1] >= 4:
    plt.figure(figsize=(10, 8))
    corr = numeric_df.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap of Numeric Features')
    plt.tight_layout()
    plt.show()
else:
    print('Not enough numeric columns for correlation heatmap')

# Pair Plot to inspect pairwise relationships between numeric features
sns.pairplot(numeric_df)
plt.suptitle('Pair Plot of Numeric Features', y=1.02)
plt.show()

# Histogram for distribution of Upvote counts
plt.figure(figsize=(8, 6))
sns.histplot(df['Upvote'], kde=True, bins=20)
plt.title('Distribution of Upvotes')
plt.xlabel('Upvote')
plt.ylabel('Frequency')
plt.show()

# Count Plot (Pie Chart alternative) for Data_Type
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Data_Type')
plt.title('Count of Datasets by Data Type')
plt.xlabel('Data Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Box Plot for Usability across different Data_Types
plt.figure(figsize=(10, 6))
sns.boxplot(x='Data_Type', y='Usability', data=df)
plt.title('Usability by Data Type')
plt.xlabel('Data Type')
plt.ylabel('Usability')
plt.xticks(rotation=45)
plt.show()
```
