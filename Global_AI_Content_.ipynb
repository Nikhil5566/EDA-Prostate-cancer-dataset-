# Visualisation

```python
plt.figure(figsize = (10, 6))
df['Year'].plot(kind='hist', bins=20, title='Year')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
plt.figure(figsize = (10, 5))
sns.histplot(df['AI Adoption Rate (%)'], bins = 30, kde = True, color = 'blue')
plt.title('Rate for AI Adoption')
plt.xlabel('AI Adoption Rate (%)')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Human-AI Collaboration Rate (%)', y='Job Loss Due to AI (%)', hue='Industry')
plt.title('Human-AI Collaboration vs Job Loss')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_countries = df['Country'].value_counts().nlargest(15).index
ax = sns.countplot(y=df[df['Country'].isin(top_countries)]['Country'],
                   order=top_countries, palette='mako')  # Fixed palette and order

plt.title(f"Most Common Countries in Dataset (Total: {len(df)} Entries)")  # Fixed f-string
plt.xlabel('Number of Entries')
plt.ylabel('Country')
plt.grid(axis='x', linestyle='--', alpha=0.3)

# Add count labels (adjusted position: width + 1 instead of +5)
for p in ax.patches:
    width = p.get_width()
    ax.text(width + 1, p.get_y() + p.get_height()/2, f'{int(width)}',
            ha='left', va='center', fontsize=10)  # Added fontsize for clarity

plt.show()
```

```python
plt.figure(figsize=(13, 6))
sns.boxplot( x="Industry", y='Market Share of AI Companies (%)', data=df, )
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
