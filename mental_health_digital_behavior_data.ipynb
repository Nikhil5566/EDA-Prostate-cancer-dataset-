# Handling Outliers:

```python
Q1 = df['notification_count'].quantile(0.25)
Q3 = df['notification_count'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_outliers = df[(df['notification_count'] >= lower_bound) & (df['notification_count'] <= upper_bound)].copy()

print('Shape before removing outliers:', df.shape)
print('Shape after removing outliers: ', df_outliers.shape)
```

# Univariate Analysis

```python
df['notification_count'].describe()
df['anxiety_level'].value_counts()
```

```python
# Histogram:

plt.figure(figsize = (15, 8))
sns.histplot(df['notification_count'])
plt.title('Histogram of Notification Count')
plt.xlabel('Notification Count')
plt.ylabel('anxity_level')
plt.show()
```

```python
# Box-Plot:

plt.figure(figsize = (15, 8))
sns.boxplot(x = df['notification_count'])
plt.title('Boxplot of notification Count')
plt.xlabel('Notification Count')
plt.show()
```

```python
# Count Plots:

plt.figure(figsize = (15, 8))
sns.countplot(x = df['anxiety_level'])
plt.title('Count plt of Anxiety Level')
plt.xlabel('Anxiety Level')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

# Bivariate Analysis:

```python
# Scatter-Plot:

plt.figure(figsize = (15, 8))
sns.scatterplot(x = df['notification_count'], y = df['anxiety_level'])
plt.title('Scatter Plot of Notification Count vs Anxiety Level')
plt.xlabel('Notification_count')
plt.ylabel('Anxiety_level')
plt.show()
```

```python
# Bar-Plot:

plt.figure(figsize = (15, 8))
sns.barplot(x = 'anxiety_level', y = 'notification_count', data=df)
plt.title('Bar Plot of Anxiety Level vs Notification Count')
plt.xlabel('Anxiety Level')
plt.ylabel('Notification Count')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Heatmap:

numerical_df = df.select_dtypes(include=np.number)
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

# Visualization:

```python
# Scatter-Plot:

plt.figure(figsize=(10, 6))
sns.scatterplot(x='sleep_hours', y='anxiety_level', data=df)
plt.title('Usage Time vs Anxiety Level')
plt.xlabel('Usage Time')
plt.ylabel('Anxiety Level')
plt.show()
```

```python
# Bar-Plot:

plt.figure(figsize=(10, 6))
sns.barplot(x='mood_score', y='anxiety_level', data=df)
plt.title('Anxiety Level by Activity Level')
plt.xlabel('Activity Level')
plt.ylabel('Anxiety Level')
plt.xticks(rotation=45) # Rotate labels if needed
plt.show()
```

```python
# Box-plot anxiety_level vs screen_time:

plt.figure(figsize=(15, 8))
sns.boxplot(x='anxiety_level', y='daily_screen_time_min', data=df)
plt.title('Distribution of Screen Time by Anxiety Level')
plt.xlabel('Anxiety Level')
plt.ylabel('Screen Time')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Violin-Plot:

plt.figure(figsize = (15, 8))
sns.violinplot(x = 'anxiety_level', y = 'focus_score', data=df)
plt.title('Age Distribution by Anxiety Level')
plt.xlabel('Anxiety Level')
plt.ylabel('Age')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
