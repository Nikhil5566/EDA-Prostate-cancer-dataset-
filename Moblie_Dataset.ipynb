# Univariate Analysis

```python
# Analyze a numerical column ('price')
print("Univariate Analysis for 'price':")
print(df['price'].describe())

# Plot the distribution of 'price'
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Analyze a categorical column ('os')
print("\nUnivariate Analysis for 'os':")
print(df['os'].value_counts())

# Plot the distribution of 'os'
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='os')
plt.title('Distribution of Operating Systems')
plt.xlabel('Operating System')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.show()

# Analyze another numerical column ('specs_score')
print("\nUnivariate Analysis for 'specs_score':")
print(df['specs_score'].describe())

# Plot the distribution of 'specs_score'
plt.figure(figsize=(10, 6))
sns.histplot(df['specs_score'], kde=True)
plt.title('Distribution of Specs Score')
plt.xlabel('Specs Score')
plt.ylabel('Frequency')
plt.show()

# Analyze a categorical column ('extra_storage')
print("\nUnivariate Analysis for 'extra_storage':")
print(df['extra_storage'].value_counts())

# Plot the distribution of 'extra_storage'
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='extra_storage')
plt.title('Distribution of Extra Storage Options')
plt.xlabel('Extra Storage')
plt.ylabel('Count')
plt.show()
```

# Bivariate Analysis

```python
# Bivariate analysis

# Check if the DataFrame is empty before proceeding
if df.empty:
    print("DataFrame is empty. Cannot perform bivariate analysis.")
else:
    # --- Numerical vs. Numerical ---

    print("\nBivariate Analysis: Numerical vs. Numerical")

    # Relationship between 'price' and 'specs_score'
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x='specs_score', y='price')
    plt.title('Price vs. Specs Score')
    plt.xlabel('Specs Score')
    plt.ylabel('Price')
    plt.show()

    # Correlation matrix for numerical columns
    # Select only numerical columns for correlation calculation
    numerical_df = df.select_dtypes(include=np.number)
    if not numerical_df.empty:
        correlation_matrix = numerical_df.corr()
        print("\nCorrelation Matrix of Numerical Features:")
        print(correlation_matrix)

        plt.figure(figsize=(8, 6))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
        plt.title('Correlation Matrix')
        plt.show()
    else:
        print("\nNo numerical columns available for correlation matrix.")


    # --- Numerical vs. Categorical ---

    print("\nBivariate Analysis: Numerical vs. Categorical")

    # Distribution of 'price' across different 'os'
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x='os', y='price')
    plt.title('Price Distribution by Operating System')
    plt.xlabel('Operating System')
    plt.ylabel('Price')
    plt.xticks(rotation=45, ha='right')
    plt.show()

    # Distribution of 'specs_score' across 'extra_storage'
    plt.figure(figsize=(8, 5))
    sns.violinplot(data=df, x='extra_storage', y='specs_score')
    plt.title('Specs Score Distribution by Extra Storage Option')
    plt.xlabel('Extra Storage')
    plt.ylabel('Specs Score')
    plt.show()

    # Grouped descriptive statistics for 'price' by 'os'
    print("\nPrice Statistics by Operating System:")
    print(df.groupby('os')['price'].describe())


    # --- Categorical vs. Categorical ---

    print("\nBivariate Analysis: Categorical vs. Categorical")

    # Count of 'os' by 'extra_storage'
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x='extra_storage', hue='os')
    plt.title('Count of OS by Extra Storage Option')
    plt.xlabel('Extra Storage')
    plt.ylabel('Count')
    plt.show()

    # Crosstabulation of 'os' and 'extra_storage'
    print("\nCrosstabulation of OS and Extra Storage:")
    print(pd.crosstab(df['os'], df['extra_storage']))
```

# Visualization

```python
# Check if the DataFrame is empty before proceeding with visualizations
if df.empty:
    print("DataFrame is empty. Cannot create visualizations.")
else:
    # Visualization Type 1: Histogram (for numerical distribution)
    print("\n--- Visualization Type 1: Histogram ---")
    plt.figure(figsize=(10, 6))
    # Using 'price' as an example numerical column
    sns.histplot(data=df, x='price', kde=True, bins=30)
    plt.title('Distribution of Price')
    plt.xlabel('Price')
    plt.ylabel('Frequency')
    plt.show()

    # Visualization Type 2: Countplot (for categorical distribution)
    print("\n--- Visualization Type 2: Countplot ---")
    plt.figure(figsize=(8, 5))
    # Using 'os' as an example categorical column
    sns.countplot(data=df, x='os')
    plt.title('Distribution of Operating Systems')
    plt.xlabel('Operating System')
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.show()

    # Visualization Type 3: Scatter Plot (for numerical vs. numerical relationship)
    print("\n--- Visualization Type 3: Scatter Plot ---")
    plt.figure(figsize=(10, 6))
    # Using 'specs_score' and 'price' as example numerical columns
    sns.scatterplot(data=df, x='specs_score', y='price', alpha=0.6)
    plt.title('Price vs. Specs Score')
    plt.xlabel('Specs Score')
    plt.ylabel('Price')
    plt.show()

    # Visualization Type 4: Box Plot (for numerical vs. categorical comparison)
    print("\n--- Visualization Type 4: Box Plot ---")
    plt.figure(figsize=(10, 6))
    # Using 'extra_storage' and 'price' as example columns
    sns.boxplot(data=df, x='extra_storage', y='price')
    plt.title('Price Distribution by Extra Storage Option')
    plt.xlabel('Extra Storage')
    plt.ylabel('Price')
    plt.show()

    # Visualization Type 5: Heatmap (for correlation matrix of numerical variables)
    print("\n--- Visualization Type 5: Heatmap ---")
    # Select only numerical columns for the heatmap
    numerical_df = df.select_dtypes(include=np.number)
    if not numerical_df.empty:
        correlation_matrix = numerical_df.corr()
        plt.figure(figsize=(8, 6))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
        plt.title('Correlation Matrix of Numerical Features')
        plt.show()
    else:
        print("No numerical columns available for heatmap.")
```
