## Advanced Data Cleaning:

### Z-score:


```python
from scipy import stats

# Calculate Z-scores for 'Typical Price (USD)'
df['price_zscore'] = np.abs(stats.zscore(df['Typical Price (USD)']))

# Define a threshold for Z-score
zscore_threshold = 3

# Filter out outliers based on the threshold
df_no_outliers_zscore = df[df['price_zscore'] < zscore_threshold].copy()

# Drop the temporary 'price_zscore' column
df_no_outliers_zscore = df_no_outliers_zscore.drop(columns=['price_zscore'])

print("Original DataFrame shape:", df.shape)
print("DataFrame shape after removing outliers (Z-score):", df_no_outliers_zscore.shape)
```

### IQR:

```python
Q1 = df['Typical Price (USD)'].quantile(0.25)
Q3 = df['Typical Price (USD)'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers_iqr = df[(df['Typical Price (USD)'] >= lower_bound) & (df['Typical Price (USD)'])].copy()

print('Original DataFrame shape', df.shape)
print('DataFrame shape after removing outliers IQR', df_no_outliers_iqr.shape)
```

# Imputing Missing Values:

### KNN:

```python
# Import the tool for KNN imputation
from sklearn.impute import KNNImputer

# Create the "tool" (imputer) and tell it how many neighbors to look at (n_neighbors)
# Looking at 2 nearest neighbors
knn_imputer = KNNImputer(n_neighbors=2)

# Identify which columns are numbers (where KNN works best)
# Use the main DataFrame 'df' instead of the undefined 'sample_df'
numerical_cols = df.select_dtypes(include=np.number).columns

# Use the "tool" to fill in the blank spaces in the number columns
# Apply imputation to the main DataFrame 'df'
df[numerical_cols] = knn_imputer.fit_transform(df[numerical_cols])
```

### Interpolation:

```python
# Identify numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns

# Loop through each numerical column and apply interpolation if it has missing values
for col in numerical_cols:
    if df[col].isnull().any(): # Check if the column has any missing values
        # Apply linear interpolation
        df[col] = df[col].interpolate(method='linear')
        print(f"Applied linear interpolation to column: {col}")

print("\nDataFrame after interpolation:")
print(df.isnull().sum()) # Check for remaining missing values
```

# Detect and handle multicollinearity (VIF)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Select only the numerical columns for VIF calculation
# We should include the encoded categorical columns as well, as they are now numerical features
# Exclude non-numeric original columns like 'Dish Name', 'Country', 'Ingredients'
# Also exclude the target variable if you have one (assuming we are preparing features for a model)

# Let's identify the columns to include. It should be all numerical columns
# after encoding, excluding any potential target variable.
# For this dataset, let's assume we're preparing features for a model
# where 'Typical Price (USD)' might be the target.
# However, VIF is typically calculated for independent variables.
# Let's calculate VIF for all numerical columns *except* the original
# 'Typical Price (USD)' if that's the variable we're trying to predict or analyze relationships with.

# Assuming we are analyzing features to predict something else,
# or just understanding relationships among features:
# Let's calculate VIF for all current numerical columns.
# You might want to drop the original 'Typical Price (USD)' if it's the target
# or if you've already created 'Price Category' from it and don't need the original.

# For demonstration, let's calculate VIF for all numerical columns present in the DataFrame
# after imputation and encoding, excluding the original 'Typical Price (USD)' if it exists
# and isn't intended as an independent variable.

# Identify numerical columns after preprocessing
numerical_cols_after_processing = df.select_dtypes(include=np.number).columns.tolist()

# If 'Typical Price (USD)' is your target variable, remove it from the feature list for VIF
# Assuming 'Typical Price (USD)' is *not* the target and we want to check multicollinearity among *all* numeric features.
# If it *is* the target, uncomment the line below to remove it from features for VIF calculation:
# if 'Typical Price (USD)' in numerical_cols_after_processing:
#    numerical_cols_after_processing.remove('Typical Price (USD)')

# If you kept 'price_zscore' from earlier, you might want to exclude it too as it's derived from 'Typical Price (USD)'
if 'price_zscore' in numerical_cols_after_processing:
    numerical_cols_after_processing.remove('price_zscore')
# Similarly for any other intermediate columns you don't want as features
if 'price_category_encoded' in numerical_cols_after_processing: # Assuming a placeholder name if you didn't use get_dummies
     numerical_cols_after_processing.remove('price_category_encoded') # Remove if using a single encoded column for price category

# Create a DataFrame with just the features for VIF calculation
X = df[numerical_cols_after_processing]

# Handle potential infinite VIF values that can occur with perfect multicollinearity
# (e.g., dummy variable trap if drop='first' wasn't used for One-Hot Encoding,
# although pd.get_dummies(drop_first=True) handles this)
# or if a feature is constant.
# We can add a small constant or check for variance.
# For simplicity, let's drop columns that have zero variance (constant values) as VIF cannot be calculated.
X = X.loc[:, X.var() != 0]

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

# Calculate VIF for each feature, handling potential errors
vif_list = []
for i in range(len(X.columns)):
    try:
        vif = variance_inflation_factor(X.values, i)
        vif_list.append(vif)
    except Exception as e:
        # Handle cases where VIF calculation fails (e.g., perfect multicollinearity)
        vif_list.append(float('inf')) # Assign infinity or a large value

vif_data["VIF"] = vif_list

print("VIF Factors:")
print(vif_data.sort_values(by="VIF", ascending=False))

# --- Handling Multicollinearity ---
# There are several ways to handle multicollinearity:
# 1. Remove one of the highly correlated variables.
# 2. Combine highly correlated variables into a single feature (e.g., averaging).
# 3. Use dimensionality reduction techniques like PCA.
# 4. Use regularization methods (Lasso or Ridge regression) which can handle multicollinearity.

# The most common approach is to remove variables with high VIF, starting with the one with the highest VIF.
# Let's define a VIF threshold (e.g., 5 or 10) and iteratively remove features.

vif_threshold = 10

# Keep track of features to drop
features_to_drop = []

# Loop and remove features until all remaining features have VIF below the threshold
# Make a copy to modify during iteration
X_filtered = X.copy()

while True:
    vif_data_filtered = pd.DataFrame()
    vif_data_filtered["feature"] = X_filtered.columns

    vif_list_filtered = []
    # Recalculate VIF for the remaining features
    for i in range(len(X_filtered.columns)):
        try:
            vif = variance_inflation_factor(X_filtered.values, i)
            vif_list_filtered.append(vif)
        except Exception as e:
            vif_list_filtered.append(float('inf'))

    vif_data_filtered["VIF"] = vif_list_filtered

    # Find the feature with the maximum VIF
    max_vif = vif_data_filtered["VIF"].max()

    # If the maximum VIF is above the threshold, remove the corresponding feature
    if max_vif > vif_threshold:
        feature_with_max_vif = vif_data_filtered.loc[vif_data_filtered["VIF"].idxmax(), "feature"]
        features_to_drop.append(feature_with_max_vif)
        print(f"Removing '{feature_with_max_vif}' with VIF = {max_vif:.2f}")
        X_filtered = X_filtered.drop(columns=[feature_with_max_vif])
    else:
        break # Stop if all VIFs are below the threshold

print("\nFeatures to drop due to high multicollinearity:", features_to_drop)
print("Remaining features after handling multicollinearity:", X_filtered.columns.tolist())

# You can now update your main DataFrame 'df' by dropping the identified features
# df = df.drop(columns=features_to_drop) # Uncomment this line to apply the changes

# Display the remaining features and their VIFs
print("\nFinal VIF Factors (after removal):")
if not X_filtered.empty:
    vif_data_final = pd.DataFrame()
    vif_data_final["feature"] = X_filtered.columns
    vif_data_final["VIF"] = [variance_inflation_factor(X_filtered.values, i) for i in range(len(X_filtered.columns))]
    print(vif_data_final.sort_values(by="VIF", ascending=False))
else:
    print("No features remaining after handling multicollinearity.")
```
