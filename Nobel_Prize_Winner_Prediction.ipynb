# Handling Outliers

```python
from scipy.stats import zscore

# Select the columns you want to check for outliers (numerical columns)
numerical_cols = df.select_dtypes(include=np.number).columns
df_numerical = df[numerical_cols]

z_scores = np.abs(zscore(df_numerical))
outlier_indices = np.where(z_scores > 3)

# outlier_indices will give you the row and column indices of the outliers.
# You can then decide how to handle them, e.g., remove or transform.
```

```python
# Select the columns you want to check for outliers (numerical columns)
numerical_cols = df.select_dtypes(include=np.number).columns

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers_iqr = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    # You can then decide how to handle outliers_iqr, e.g., remove or transform.
    # print(f"Outliers in column '{col}':\n{outliers_iqr}")
```

```python
# Histograms for Numerical Columns

# Select the numerical columns, excluding 'year'
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
if 'year' in numerical_cols:
      numerical_cols.remove('year')

# Create histograms for each selected numerical column
if numerical_cols: # Check if there are any numerical columns left
    plt.figure(figsize=(15, 10)) # Adjust the figure size as needed
    num_cols = len(numerical_cols)
    n_cols_plot = 3
    n_rows_plot = num_cols // n_cols_plot + (num_cols % n_cols_plot > 0)

for i, col in enumerate(numerical_cols):
      plt.subplot(n_rows_plot, n_cols_plot, i + 1) # Arrange plots in a grid
      df[col].hist(bins=20) # You can adjust the number of bins
      plt.title(f'Histogram of {col}')
      plt.xlabel(col)
      plt.ylabel('Frequency')

      plt.tight_layout() # Adjust spacing between plots
      plt.show()
else:
    print("No other numerical columns found for histograms after excluding 'year'.")
```

```python
# Bar Plots for Categorical Columns

# List of potential categorical columns you might have
categorical_columns_to_plot = ['category', 'gender', 'country'] # Replace with actual column names

for col in categorical_columns_to_plot:
  if col in df.columns and df[col].dtype == 'object': # Check if column exists and is of object type (often indicates categorical)
    plt.figure(figsize=(10, 6))
    # For columns with many unique values (like country), you might want to show only the top N
    if df[col].nunique() > 20: # Example threshold, adjust as needed
      top_n = 20
      top_categories = df[col].value_counts().nlargest(top_n).index
      sns.countplot(y=df[col], order=top_categories)
      plt.title(f'Top {top_n} Distribution of {col}')
    else:
      sns.countplot(y=df[col], order=df[col].value_counts().index)
      plt.title(f'Distribution of {col}')

    # These lines should be indented to be part of the 'if col in df.columns and df[col].dtype == 'object':' block
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()
  elif col in df.columns:
    print(f"Column '{col}' exists but is not of object type. Consider if it should be treated as categorical.")
  else:
    print(f"Column '{col}' not found.")
```

# Visualization

```python
# --- Visualization 1: Histogram of 'year' ---
# This shows the distribution of Nobel Prizes over the years.
if 'awardYear' in df.columns:
    plt.figure(figsize=(10, 6))
    df['awardYear'].hist(bins=30) # Adjust bins as needed
    plt.title('Distribution of Nobel Prizes by Year')
    plt.xlabel('Year')
    plt.ylabel('Number of Prizes')
    plt.show()
```

```python
# --- Visualization 2: Bar Plot of 'category' ---
# This shows the number of prizes awarded in each category.
if 'category' in df.columns:
    plt.figure(figsize=(12, 7))
    df['category'].value_counts().plot(kind='bar')
    plt.title('Number of Nobel Prizes by Category')
    plt.xlabel('Category')
    plt.ylabel('Number of Prizes')
    plt.xticks(rotation=45, ha='right') # Rotate labels for readability
    plt.tight_layout() # Adjust layout
    plt.show()
```

```python
# --- Visualization 3: Count Plot of 'gender' (using Seaborn) ---
# This shows the distribution of Nobel laureates by gender.
if 'gender' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(x='gender', data=df)
    plt.title('Distribution of Nobel Laureates by Gender')
    plt.xlabel('Gender')
    plt.ylabel('Count')
    plt.show()
```

```python
numerical_cols_for_scatter = df.select_dtypes(include=np.number).columns.tolist()
if 'awardYear' in numerical_cols_for_scatter:
    numerical_cols_for_scatter.remove('awardYear')

if numerical_cols_for_scatter:
    # Use the first available numerical column other than year
    numerical_col_to_plot = numerical_cols_for_scatter[0]
    plt.figure(figsize=(10, 6))
    plt.scatter(df['awardYear'], df[numerical_col_to_plot])
    plt.title(f'Scatter Plot of {numerical_col_to_plot} vs. Year')
    plt.xlabel('Year')
    plt.ylabel(numerical_col_to_plot)
    plt.show()
elif 'awardYear' in df.columns:
     # If no other numerical column, scatter plot year against row index
     plt.figure(figsize=(10, 6))
     plt.scatter(df['awardYear'], df.index)
     plt.title('Scatter Plot of Row Index vs. Year')
     plt.xlabel('Year')
     plt.ylabel('Row Index')
     plt.show()
else:
    print("Not enough numerical columns for Scatter Plot.")
```

```python
# --- Visualization 5: Box Plot of 'year' by 'category' ---

# This shows the distribution of the years Nobel Prizes were awarded,
# separated by category, to see if certain categories were awarded more often
# in particular periods.
if 'awardYear' in df.columns and 'category' in df.columns:
    plt.figure(figsize=(14, 8)) # Adjust figure size
    sns.boxplot(x='category', y='awardYear', data=df)
    plt.title('Distribution of Nobel Prize Years by Category')
    plt.xlabel('Category')
    plt.ylabel('Year')
    plt.xticks(rotation=45, ha='right') # Rotate labels for readability
    plt.tight_layout() # Adjust layout
    plt.show()
else:
    print("Required columns ('awardYear', 'category') not found for Box Plot.")
```

```python
# Heatmap of Correlation Matrix:

plt.figure(figsize=(10, 8))
# Select only numerical columns before calculating the correlation matrix
numerical_df = df.select_dtypes(include=np.number)
correlation_matrix = numerical_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
