```python
sns.barplot(x=df['Gender'],y=df['Average Screen Time (hours)'])
```

```python
sns.boxplot(x=df['Screen Time Type'],y=df['Average Screen Time (hours)'])
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Age"], y=df["Average Screen Time (hours)"], palette="coolwarm")
plt.title("Average screen time as per the age group", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Screen Time")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
