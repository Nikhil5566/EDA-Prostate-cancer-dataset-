# Handling Outlires

```python
sns.boxplot(df['life_expectancy'])
plt.show()

Q1 = df['life_expectancy'].quantile(0.25)
Q3 = df['life_expectancy'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['life_expectancy'] < lower_bound) | (df['life_expectancy'] > upper_bound)]
```

# Univariate analysis:


```python
# Numerical

# Descriptive statistics
print(df['life_expectancy'].describe())

# Histogram
plt.hist(df['life_expectancy'])
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.title('Distribution of Life Expectancy')
plt.show()

# Box plot
sns.boxplot(df['life_expectancy'])
plt.title('Box Plot of Life Expectancy')
plt.show()
```

```python
# Categorical

# Frequency table
print(df['governance_quality_index'].value_counts())

# Bar chart
sns.countplot(df['governance_quality_index'])
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of Categorical Variable')
plt.show()
```

# Bivariate analysis:


```python
# Numerical vs. Numerical

# Scatter plot
plt.scatter(df['child_mortality'], df['school_enrollment_secondary'])
plt.xlabel('child_mortality')
plt.ylabel('school_enrollment_secondary')
plt.title('Scatter Plot of Variable 1 vs. Variable 2')
plt.show()

# Correlation analysis
correlation = df['child_mortality'].corr(df['school_enrollment_secondary'])
print(f"Correlation coefficient: {correlation}")
```

```python
# Categorical vs. Categorical

# Cross-tabulation
cross_tab = pd.crosstab(df['currency_unit'], df['gdp_usd'])
print(cross_tab)

# Chi-square test
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(cross_tab)
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
```

```python
# Numerical vs. Categorical

# Box plots
sns.boxplot(x='electricity_access_pct', y='years_since_century', data=df)
plt.xlabel('Categorical Variable')
plt.ylabel('Numerical Variable')
plt.title('Box Plot of Numerical Variable by Categorical Variable')
plt.show()
```

# Visualization

```python
#Scatter Plot

# Assuming 'continent' is a categorical column in your DataFrame
continent_counts = df['year'].value_counts()
plt.pie(continent_counts, labels=continent_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Continents')
plt.show()
```

```python
sns.violinplot(y=df['life_expectancy'])
plt.ylabel('Life Expectancy')
plt.title('Distribution of Life Expectancy')
plt.show()
```

```python
sns.countplot(df['currency_unit'])
plt.xlabel('Currency Unit')
plt.ylabel('Count')
plt.title('Bar Chart of Currency Units')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
plt.figure(figsize=(20, 8))
# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=np.number)
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
