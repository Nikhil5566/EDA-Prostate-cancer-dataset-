# Univariate Analysis

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Media Coverage Fair', data=df)
plt.title('Distribution of Media Coverage Fair')
plt.xlabel('Media Coverage Fair')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.countplot(y='State', data=df, order=df['State'].value_counts().index)
plt.title('Distribution of CategoricalColumn')
plt.xlabel('Count')
plt.ylabel('CategoricalColumn')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Election Year', kde=True)
plt.title('Distribution of NumericalColumn')
plt.xlabel('NumericalColumn')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Expert ID')
plt.title('Box Plot of NumericalColumn')
plt.ylabel('NumericalColumn')
plt.show()
```

```python
categorical_counts = df['Leaders Selected Candidates'].value_counts()
print(categorical_counts)
```

# Bivariate Analysis

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Leaders Selected Candidates', y='Expert ID', data=df)
plt.title('NumericalColumn by CategoricalColumn')
plt.xlabel('CategoricalColumn')
plt.ylabel('NumericalColumn')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Media Coverage Index', y='Vote Count Index', data=df)
plt.title('Media Coverage Index vs Vote Count Index')
plt.xlabel('Media Coverage Index')
plt.ylabel('Vote Count Index')
plt.show()
```

```python
# Create a cross-tabulation
ct = pd.crosstab(df['Candidate Supported'], df['Citizens Not Found in Register'])

# Plot a stacked bar plot
ct.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Relationship between Candidate Supported and Citizens not found in register')
plt.xlabel('Candidate Supported')
plt.ylabel('Count')
plt.show()

# Plot a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(ct, annot=True, fmt='d', cmap='Blues')
plt.title('Heatmap of Candidate Supported and Citizens Not Found in Register')
plt.xlabel('Citizens Not Found in Register')
plt.ylabel('Candidate Supported')
plt.show()
```

```python
# Group by categorical column and calculate the mean of the numerical column
grouped_data = df.groupby('Protests Violent')['Electoral Authorities Index'].mean().reset_index()

# Plot a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Protests Violent', y='Electoral Authorities Index', data=grouped_data)
plt.title('Mean of Electoral Authorities Index by CategoricalColumn')
plt.xlabel('Protests Violent')
plt.ylabel('Mean Electoral Authorities Index')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels if they are long
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# Select only numerical columns
numerical_df = df.select_dtypes(include=np.number)

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(15, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
