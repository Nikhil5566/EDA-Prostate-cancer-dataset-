```python
import numpy as n
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings('ignore')
```

```python
missing_values = df.isnull().sum()

duplicate_rows = df.duplicated().sum()

missing_values, duplicate_rows
```

```python
plt.figure(figsize = (10, 6))
plt.bar(x = df['Client_Region'], height = df['Payment_Method'], color = 'grey')
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.countplot(x=df['Platform'], palette="viridis")
plt.title("Working platform", fontsize=14)
plt.xlabel("Website to find work")
plt.ylabel("Amount")

plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["Freelancer_ID"], y=df["Job_Category"], palette="coolwarm")
plt.title("Job Work", fontsize=14)
plt.xlabel("ID`s")
plt.ylabel("Type of Work")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=1)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
