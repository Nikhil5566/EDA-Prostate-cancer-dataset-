```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
plt.figure(figsize = (15,8))   # Increasing the size of the bar-chart:
plt.bar(x = df['Age'], height = df['Sell Price'], color = 'blue')

plt.title('Car sells as per age', fontsize = 16, fontweight = 'bold')
plt.xlabel('Age', fontsize = 16, fontweight = 'bold')
plt.ylabel('Price', fontsize = 16, fontweight = 'bold')

plt.show()
#plt.savefig('Used Cars bar.png')
```

```python
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Gender'].value_counts()
    for x_label, grp in df.groupby('Cus. Location')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Cus. Location')
_ = plt.ylabel('Gender')
```

```python
df.groupby('Cus. Location').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df.plot(kind='scatter', x='Age', y='Sell Price', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
