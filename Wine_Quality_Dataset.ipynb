```python
# To check which columns has duplicates:

duplicate_columns = [col for col in df.columns if df[col].duplicated().any()]
print("Columns with duplicate values:", duplicate_columns)
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

# Handling Outliners:

```python
# Outliers can skew your analysis. You can use box plots to identify them and then decide whether to remove or cap them:

# Example using 'total_sulfur_dioxide' column
sns.boxplot(df['total sulfur dioxide'])
plt.show()

# Remove outliers using IQR (Interquartile Range)
Q1 = df['total sulfur dioxide'].quantile(0.25)
Q3 = df['total sulfur dioxide'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['total sulfur dioxide'] >= Q1 - 1.5 * IQR) & (df['total sulfur dioxide'] <= Q3 + 1.5 * IQR)]
```

#  Univariate analysis:

```python
# 4. Exploratory Data Analysis (EDA):
# Univariate analysis: Visualize the distribution of individual variables:

# Histogram for 'alcohol'
plt.hist(df['alcohol'])
plt.xlabel('Alcohol')
plt.ylabel('Frequency')
plt.show()

# Box plot for 'pH'
sns.boxplot(df['pH'])
plt.show()
```

# Bivariate analysis:

```python
# Explore relationships between variables:


# Scatter plot between 'alcohol' and 'quality':

plt.scatter(df['alcohol'], df['quality'])
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.show()

# Correlation matrix
plt.figure(figsize = (10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

# Feature Engineering:

```python
# 5. Feature Engineering: Creating new features: Combine existing features to create new ones that might be more informative. For example, you could create a feature representing the ratio of two existing features:

# Create a new feature 'total_acidity'
df['total acidity'] = df['fixed acidity'] + df['volatile acidity']
```

```python
# 2. Advanced EDA
# Distribution Analysis: Use histograms and density plots to visualize the distribution of each feature:

# Histogram for 'alcohol'
sns.histplot(df['alcohol'], kde=True)  # kde=True adds a density curve
plt.show()

# Density plot for 'pH'
sns.kdeplot(df['pH'])
plt.show()
```

```python
# Relationship Exploration: Use scatter plots, pair plots, and other visualization techniques to explore relationships:
# Pair plot for selected features
sns.pairplot(df[['alcohol', 'pH', 'quality']])
plt.show()

# Joint plot with regression line
sns.jointplot(x='alcohol', y='quality', data=df, kind='reg')
plt.show()
```

```python
# Group Comparisons: Use box plots, violin plots, or bar charts to compare feature distributions across groups:
# Box plot of 'alcohol' by 'quality'
sns.boxplot(x='quality', y='alcohol', data=df)
plt.show()

# Violin plot of 'pH' by 'quality'
sns.violinplot(x='quality', y='pH', data=df)
plt.show()
```

# Visualization

```python
#  Histogram of Alcohol Content:

plt.figure(figsize=(8, 6))
sns.histplot(df['alcohol'], kde=True)
plt.title('Distribution of Alcohol Content')
plt.xlabel('Alcohol (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot of Quality vs. Volatile Acidity:

plt.figure(figsize=(8, 6))
sns.boxplot(x='quality', y='volatile acidity', data=df)
plt.title('Quality vs. Volatile Acidity')
plt.xlabel('Quality')
plt.ylabel('Volatile Acidity')
plt.show()
```

```python
# Scatter Plot of Alcohol vs. Density:

plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='density', data=df, hue='quality')
plt.title('Alcohol vs. Density')
plt.xlabel('Alcohol (%)')
plt.ylabel('Density')
plt.show()
```

```python
# Violin Plot of pH by Quality:

plt.figure(figsize=(8, 6))
sns.violinplot(x='quality', y='pH', data=df)
plt.title('pH by Quality')
plt.xlabel('Quality')
plt.ylabel('pH')
plt.show()
```

```python
# Heatmap of Correlation Matrix:

plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
