```python
plt.figure(figsize=(8, 5))
sns.countplot(x=df['duration'], palette="viridis")
plt.title("Distribution of Passenger Numbers", fontsize=14)
plt.xlabel("Number of Passengers")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x=df['passenger_count'], palette="viridis")
plt.title("Distribution of Passenger Numbers", fontsize=14)
plt.xlabel("Number of Passengers")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["payment_type"], y=df["fare_amount"], palette="coolwarm")
plt.title("Payment Type vs. Fare Amount", fontsize=14)
plt.xlabel("Payment Type")
plt.ylabel("Fare Amount ($)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
