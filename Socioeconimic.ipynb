# Visualization

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.title('Sex Distribution')
sns.countplot(x=df['Sex'], palette='viridis')
plt.subplot(1,2,2)
plt.title('Education Distribution')
sns.countplot(x=df['Education'])
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.title('Sex vs Education')
sns.countplot(x=df['Sex'], hue=df['Education'])
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.title('Sex vs Marital Status')
sns.countplot(x=df['Sex'], hue=df['Marital status'], palette='Reds')
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.title('Occupation vs Education')
sns.countplot(x=df['Occupation'], hue=df['Education'])
plt.xticks(rotation=80)
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.title('Settlement size vs Sex')
sns.countplot(x=df['Settlement size'], hue=df['Sex'], palette='magma')
plt.show()
```

```python
# Heatmap:

plt.figure(figsize=(9,6))
plt.title('Correlation of Numerical Columns')
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='mako')
plt.show()
```
