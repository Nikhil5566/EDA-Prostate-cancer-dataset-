```python
df = df.drop(columns=['InvoiceDate'])
```

```python
df = df.drop(columns=['CustomerID'])
```

```python
df = df.drop(columns=['Description'])
```

```python
df.info()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=df['StockCode'], palette="viridis")
plt.title("StockCode", fontsize=14)
plt.xlabel("Company Names")
plt.ylabel("Comapany Codes")
plt.xticks(rotation= 70)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=df['Quantity'], palette="viridis")
plt.title("Distribution of Passenger Numbers", fontsize=14)
plt.xlabel("X-Axis")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["UnitPrice"], y=df["Country"], palette="coolwarm")
plt.title("Country vs Prices", fontsize=14)
plt.xlabel("Prices")
plt.ylabel("Country")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=1)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
