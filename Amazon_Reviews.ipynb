# Handling Outliers

```python
import scipy.stats as stats


z_scores = np.abs(stats.zscore(df['Score']))
outlier_indices_zscore = np.where(z_scores > 3)
print("Indices of outliers (Z-Score > 3):", outlier_indices_zscore)

Q1 = df['Score'].quantile(0.25)
Q3 = df['Score'].quantile(0.75)
IQR = Q3 - Q1

lower_bound_iqr = Q1 - 1.5 * IQR
upper_bound_iqr = Q3 + 1.5 * IQR

outliers_iqr = df[(df['Score'] < lower_bound_iqr) | (df['Score'] > upper_bound_iqr)]
print("\nOutlier data points (IQR method):\n", outliers_iqr)
```

# Univariate Analysis

```python
# Display the distribution of the 'Score' column
plt.figure(figsize=(8, 6))
sns.histplot(df['Score'], kde=True, bins=5)
plt.title('Distribution of Score')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()

# Display value counts for 'Score'
print("\nValue counts for Score:\n", df['Score'].value_counts())

# Display descriptive statistics for 'Score'
print("\nDescriptive statistics for Score:\n", df['Score'].describe())
```

# Bivariate Analysis

```python
df['HelpfulnessRatio'] = df['HelpfulnessNumerator'] / (df['HelpfulnessDenominator'] + 1e-9)

# Relationship between Score and Helpfulness Ratio
plt.figure(figsize=(10, 6))
sns.boxplot(x='Score', y='HelpfulnessRatio', data=df)
plt.title('Helpfulness Ratio vs. Score')
plt.xlabel('Score')
plt.ylabel('Helpfulness Ratio')
plt.show()

# Let's also look at the relationship between Score and the total number of helpfulness votes
df['TotalHelpfulnessVotes'] = df['HelpfulnessNumerator'] + df['HelpfulnessDenominator']

plt.figure(figsize=(10, 6))
sns.boxplot(x='Score', y='TotalHelpfulnessVotes', data=df)
plt.title('Total Helpfulness Votes vs. Score')
plt.xlabel('Score')
plt.ylabel('Total Helpfulness Votes')
plt.yscale('log')
plt.show()
```

# Visualization

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['ProductId'].astype(str), bins=50)
plt.title('Distribution of Product IDs')
plt.xlabel('Product ID')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Score', data=df)
plt.title('Count of Reviews by Score')
plt.xlabel('Score')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
# Scatterplot of Helpfulness Ratio vs. Total Helpfulness Votes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalHelpfulnessVotes', y='HelpfulnessRatio', data=df, alpha=0.5)
plt.title('Helpfulness Ratio vs. Total Helpfulness Votes')
plt.xlabel('Total Helpfulness Votes')
plt.ylabel('Helpfulness Ratio')
plt.xscale('log') # Use a log scale due to potential skewness
plt.show()
```

```python
# Scatterplot of Score vs. Helpfulness Denominator
plt.figure(figsize=(10, 6))
sns.scatterplot(x='HelpfulnessDenominator', y='Score', data=df, alpha=0.5)
plt.title('Score vs. Helpfulness Denominator')
plt.xlabel('Helpfulness Denominator')
plt.ylabel('Score')
plt.xscale('log') # Use a log scale due to potential skewness
plt.show()
```

```python
# Distribution plot of Helpfulness Denominator
plt.figure(figsize=(8, 6))
sns.histplot(df['HelpfulnessDenominator'], kde=True)
plt.title('Distribution of Helpfulness Denominator')
plt.xlabel('Helpfulness Denominator')
plt.ylabel('Frequency')
plt.xscale('log') # Use a log scale due to potential skewness
plt.show()
```

```python
# Heatmap of Correlation Matrix:

plt.figure(figsize=(10, 8))
# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=np.number)
correlation_matrix = numerical_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
