```python
# Boxplot: Categorical vs. Numerical Example:

plt.figure(figsize=(12, 6))
sns.boxplot(x=df['Anxiety_Score'], y=df['Stress_Level'], palette='coolwarm')
plt.title('Stress Level Distribution by Gender')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x=df['Stress_Level'], palette="viridis")
plt.title("Bar Chart for Stree level meter", fontsize=14)
plt.xlabel(" Stress Meter")
plt.ylabel("Age")

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=df["Gender"], y=df["Stress_Level"], palette="coolwarm")
plt.title("Stress Level Distribution by Gender", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Stress Level")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=1)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
