# Handling Outliers

```python
Q1 = df['Job Satisfaction Score (1-10)'].quantile(0.25)
Q3 = df['Job Satisfaction Score (1-10)'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 - 1.5 * IQR

df_outliers = df[(df['Job Satisfaction Score (1-10)'] >= lower_bound) | (df['Job Satisfaction Score (1-10)'] <= upper_bound)]
print('Total Outliers after removing: ', df_outliers.shape[0])
```

```python
# Visualize outliers using a box plot for 'Job Satisfaction Score (1-10)'
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Job Satisfaction Score (1-10)'])
plt.title('Box Plot to Visualize Outliers in Job Satisfaction Scores')
plt.xlabel('Job Satisfaction Score')
plt.show()
```

# Univariate Analysis:

```python
# Univariate Analysis for 'Job Satisfaction Score (1-10)'
plt.figure(figsize=(12, 5))

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(df['Job Satisfaction Score (1-10)'], kde=True)
plt.title('Distribution of Job Satisfaction Scores')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Frequency')

# Box Plot
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Job Satisfaction Score (1-10)'])
plt.title('Box Plot of Job Satisfaction Scores')
plt.xlabel('Job Satisfaction Score')

plt.tight_layout()
plt.show()
```

# Bivariate Analysis:

```python
# Bivariate Analysis: Job Satisfaction Score vs. Salary in USD
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Salary (Annual)', y='Job Satisfaction Score (1-10)', data=df)
plt.title('Job Satisfaction Score vs. Salary in USD')
plt.xlabel('Salary in USD')
plt.ylabel('Job Satisfaction Score (1-10)')
plt.show()
```
