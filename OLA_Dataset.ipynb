# Visaulization

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['count'], kde=True, color='blue')
plt.title('Distribution of Ride Requests (Count)')
plt.xlabel('Count')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Count plot for season (treated as categorical even though stored as numeric)
plt.figure(figsize=(8, 5))
sns.countplot(x='season', data=df, palette='viridis')
plt.title('Ride Requests by Season')
plt.xlabel('Season')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Box plot for temperature with respect to ride count
plt.figure(figsize=(8, 5))
sns.boxplot(x='temp', y='count', data=df, palette='coolwarm')
plt.title('Box Plot of Ride Count by Temperature')
plt.xlabel('Temperature')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Pair plot of a few selected features to observe pairwise relationships
cols_to_plot = ['temp', 'humidity', 'windspeed', 'count']
sns.pairplot(df[cols_to_plot], diag_kind='kde', plot_kws={'alpha':0.5})
plt.suptitle('Pair Plot of Selected Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
