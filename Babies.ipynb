# Visualization

```python
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols].hist(figsize=(14,12), bins=30)
plt.tight_layout()

plt.show()
```

```python
# 1. Histogram for 'bwt' (Birth Weight)
plt.figure(figsize=(8,5))
sns.histplot(df['bwt'], kde=True, bins=30)
plt.title('Distribution of Birth Weight')
plt.xlabel('Birth Weight')
plt.ylabel('Frequency')
plt.savefig('birth_weight_histogram.png')
plt.show()
```

```python
# 2. Box Plot for 'bwt' vs 'smoke' (smoking status)
plt.figure(figsize=(8,5))
sns.boxplot(x='smoke', y='bwt', data=df)
plt.title('Birth Weight by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Birth Weight')
plt.savefig('bwt_smoke_boxplot.png')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
