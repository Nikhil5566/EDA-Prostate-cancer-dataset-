```python
# Renaming the columns names by removing the spaces in between the its columns name:

df = df.rename(columns={'Water Treatment Method': 'Water_Treatment_Method'})
```

```python
df['Water_Treatment_Method'].fillna(df['Water_Treatment_Method'].mode()[0], inplace=True)
```

```python
sns.scatterplot(x='Year', y = 'Lead Concentration (µg/L)',  data = df)
```

# Visualization

```python
numeric_cols = df.select_dtypes(include=np.number).columns

df[numeric_cols].hist(figsize=(14,12), bins=30)
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=df.groupby('Country')['Access to Clean Water (% of Population)'].mean().reset_index(),
            x='Access to Clean Water (% of Population)', y='Country')
plt.title("Average Access to Clean Water by Country")
```

```python
sns.boxplot(data=df, x='Water Source Type', y='Contaminant Level (ppm)')
plt.xticks(rotation=45)
plt.title('Contaminant Level by Water Source')
```

```python
plt.figure(figsize=(22, 10))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
