# Visualization

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Vehicle Type Involved', data=df, palette='viridis')
plt.title('Vehicle Type Involved Distribution (0 = Not Churned, 1 = Churned)')
plt.xlabel('Vehicle Type Involved')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Driver Age'], bins=30, kde=True, color='blue')
plt.title('Driver Age Distribution of Customers')
plt.xlabel('Driver Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df['Year'], color='green')
plt.title('Boxplot of Year')
plt.ylabel('Year')
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
