```python
# From your df.info(), numerical columns are float64
numerical_cols = ['Vitamin C', 'Vitamin B11', 'Sodium', 'Calcium', 'Carbohydrates',
                  'Iron', 'Caloric Value', 'Sugars', 'Dietary Fiber', 'Fat', 'Protein']

# Categorical columns are object type
categorical_cols = ['food', 'food normalized']

# Verify
print("Numerical columns:", numerical_cols)
print("Categorical columns:", categorical_cols)
```

# Visualization

```python
# Get numerical columns:
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Create a grid of histograms
fig, axes = plt.subplots(nrows=len(numerical_cols), ncols=1, figsize=(10, 5*len(numerical_cols)))

for col, ax in zip(numerical_cols, axes):
    ax.hist(df[col], bins=30, color='skyblue', edgecolor='black')
    ax.set_title(f'Distribution of {col}', fontsize=12)
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')

plt.tight_layout()  # Prevent overlapping labels
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Get categorical columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

# Create subplots for each categorical-numerical pair
for cat_col in categorical_cols:
    for num_col in numerical_cols[:3]:  # Just showing first 3 numerical for demo
        plt.figure(figsize=(10, 5))

        # Boxplot shows distribution stats
        sns.boxplot(x=cat_col, y=num_col, data=df)

        # Add swarmplot to show individual points
        sns.swarmplot(x=cat_col, y=num_col, data=df, color='black', alpha=0.5)

        plt.title(f'{num_col} by {cat_col}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

```python
# Top 10 highest coleries food:

top_calories = df[['food_normalized', 'Caloric Value']].sort_values(by='Caloric Value', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(data=top_calories, y='food_normalized', x='Caloric Value', palette='Reds_r')
plt.title('Top 10 Highest Calorie Foods')
plt.xlabel('Calories (kcal)')
plt.ylabel('Food Item')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
