# Handling Outliers

```python
Q1 = df['years_experience'].quantile(0.25)
Q3 = df['years_experience'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q1 + 1.5 * IQR

df_outliers = df[(df['years_experience'] >= lower_bound) & (df['years_experience'] <= upper_bound)].copy()


print('Original number of rows:', len(df))
print('Number of rows after removing outliers:', len(df_outliers))
```

```python
# Creating a box plot of the 'years_experience' column:

# Boxplot after before removing the outliers:
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['years_experience'])
plt.title('Box Plot of Years of Experience')
plt.xlabel('Years of Experience')
plt.show()

# Box-plot after removing outliers:
plt.figure(figsize=(8, 6))
sns.boxplot(x=df_outliers['years_experience'])
plt.title('Box Plot of Years of Experience (After Outlier Removal)')
plt.xlabel('Years of Experience')
plt.show()
```

# Univariate Analysis:

```python
# --- Univariate Analysis for Numerical Columns ---

numerical_cols = ['years_experience']

print("--- Univariate Analysis for Numerical Columns ---")
for col in numerical_cols:
    print(f"\nAnalyzing: {col}")

    # Basic statistics
    print("Basic Statistics:")
    print(df[col].describe())

    # Check for missing values
    print(f"\nMissing values: {df[col].isnull().sum()}")

    # Visualize the distribution (Histogram and Box Plot)
    plt.figure(figsize=(12, 5))

    # Histogram
    plt.subplot(1, 2, 1)
    # Full-Form: kde=True to show Kernel Density Estimate:
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

    # Box Plot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
    plt.xlabel(col)

    plt.tight_layout()
    plt.show()


# --- Univariate Analysis for Categorical Columns ---

# Corrected the typo in 'employement_type' to 'employment_type'
categorical_cols = ['job_id', 'job_title', 'salary_currency', 'experience_level', 'employment_type']

print("\n--- Univariate Analysis for Categorical Columns ---")
for col in categorical_cols:
    print(f"\nAnalyzing: {col}")

    # Value counts (frequency of each category)
    print("Value Counts:")
    print(df[col].value_counts())

    # Check for missing values
    print(f"\nMissing values: {df[col].isnull().sum()}")

    # Visualize the distribution (Count Plot):
    plt.figure(figsize=(10, 6))
    sns.countplot(y=df[col], order = df[col].value_counts().index, palette='viridis')
    plt.title(f'Count of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()

print("\n--- Univariate Analysis for Job Title (Top 10) ---")
top_n = 10
print(f"Value Counts (Top {top_n}):")
print(df['job_title'].value_counts().head(top_n))

plt.figure(figsize=(10, 6))
sns.countplot(y=df['job_title'], order = df['job_title'].value_counts().head(top_n).index, palette='viridis')
plt.title(f'Count of Top {top_n} Job Titles')
plt.xlabel('Count')
plt.ylabel('Job Title')
plt.show()

# --- Univariate Analysis for Date Column ---

print("\n--- Univariate Analysis for Job Posting Date ---")
# Assuming 'job_posting_date' is in datetime format or can be converted
try:
    df['job_posting_date'] = pd.to_datetime(df['job_posting_date'])
    print("Basic Statistics for Date:")
    print(df['job_posting_date'].describe(datetime_is_numeric=True)) # Use datetime_is_numeric for date stats

    # Visualize distribution over time (e.g., count of postings per month/year)
    df['post_month_year'] = df['job_posting_date'].dt.to_period('M') # Group by month and year
    monthly_counts = df['post_month_year'].value_counts().sort_index()

    plt.figure(figsize=(12, 6))
    monthly_counts.plot(kind='bar')
    plt.title('Number of Job Postings Over Time (Monthly)')
    plt.xlabel('Month/Year')
    plt.ylabel('Number of Postings')
    plt.xticks(rotation=45, ha='right') # Rotate labels for readability
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"\nCould not perform analysis on 'job_posting_date'. Error: {e}")
    print("Please ensure the 'job_posting_date' column exists and is in a datetime format or can be converted.")
```

# Bivariate Analysis:

```python
categorical_cols_for_numerical_comparison = ['experience_level', 'employment_type']

print("\n--- Bivariate Analysis: Numerical vs. Categorical ---")
numerical_col_to_analyze = 'years_experience'

for cat_col in categorical_cols_for_numerical_comparison:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=cat_col, y=numerical_col_to_analyze, data=df, palette='viridis')
    plt.title(f'{numerical_col_to_analyze} by {cat_col}')
    plt.xlabel(cat_col)
    plt.ylabel(numerical_col_to_analyze)
    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels if needed
    plt.tight_layout()
    plt.show()

# Using Violin Plots (similar to box plots but show density)
for cat_col in categorical_cols_for_numerical_comparison:
    plt.figure(figsize=(10, 6))
    sns.violinplot(x=cat_col, y=numerical_col_to_analyze, data=df, palette='viridis')
    plt.title(f'{numerical_col_to_analyze} Distribution by {cat_col}')
    plt.xlabel(cat_col)
    plt.ylabel(numerical_col_to_analyze)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# --- Bivariate Analysis: Categorical vs. Categorical ---

categorical_cols_to_compare = ['experience_level', 'employment_type']

print("\n--- Bivariate Analysis: Categorical vs. Categorical ---")

# Using Count Plots or Heatmaps of Cross-Tabulations
for i in range(len(categorical_cols_to_compare)):
    for j in range(i + 1, len(categorical_cols_to_compare)):
        col1 = categorical_cols_to_compare[i]
        col2 = categorical_cols_to_compare[j]

        print(f"\nAnalyzing relationship between {col1} and {col2}:")

        # Create a cross-tabulation
        crosstab = pd.crosstab(df[col1], df[col2])
        print("Cross-tabulation:")
        print(crosstab)

        # Visualize using a heatmap
        plt.figure(figsize=(12, 8))
        sns.heatmap(crosstab, annot=True, fmt='d', cmap='Blues')
        plt.title(f'Relationship between {col1} and {col2}')
        plt.xlabel(col2)
        plt.ylabel(col1)
        plt.show()

# Analyzing 'salary_range' vs other categorical features can be very insightful:
print("\n--- Bivariate Analysis: Salary Range vs. Other Categorical Features ---")
# Select categorical columns that exist in the DataFrame
categorical_cols_vs_salary = ['experience_level', 'employment_type']

for cat_col in categorical_cols_vs_salary:
    print(f"\nAnalyzing relationship between {cat_col} and salary_range:")
    if 'salary_range' in df.columns:
        crosstab = pd.crosstab(df[cat_col], df['salary_range'])
        print("Cross-tabulation:")
        print(crosstab)

        plt.figure(figsize=(12, 8))
        sns.heatmap(crosstab, annot=True, fmt='d', cmap='Blues')
        plt.title(f'Relationship between {cat_col} and salary_range')
        plt.xlabel('Salary Range')
        plt.ylabel(cat_col)
        plt.show()
    else:
        print(f"Column 'salary_range' not found in the DataFrame.")
```
