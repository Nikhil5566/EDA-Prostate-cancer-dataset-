# Visualization

```python
# Selecting numeric columns for further analysis
numeric_df = df.select_dtypes(include=[np.number])
print('Numeric columns:', list(numeric_df.columns))

# Count plot for categorical variable: Brand
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='Brand', order=df['Brand'].value_counts().index, palette='viridis')
plt.title('Count of Sales by Brand')
plt.show()
```

```python
# Box plot for Price grouped by Brand
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Brand', y='Price', palette='Set2')
plt.title('Price Distribution across Brands')
plt.xticks(rotation=45)
plt.show()
```

```python
# Additional Visualization: Boxen plot for Quantity Sold by Region
plt.figure(figsize=(12,6))
sns.boxenplot(data=df, x='Region', y='Quantity Sold', palette='coolwarm')
plt.title('Quantity Sold by Region')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
