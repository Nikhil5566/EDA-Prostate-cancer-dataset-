# Visualization

```python
plt.figure(figsize = (10, 5))
sns.countplot(x = 'continent', data = df)
plt.title('Countries')

plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.histplot(df['new_deaths'], bins = 30, kde = True, color = 'blue')
plt.title('Deaths')

plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.boxplot(x = 'total_deaths', y = 'new_deaths', data=df)
plt.title('Motor UPDRS by sex')

plt.show()
```

```python
plt.figure(figsize=(18, 9))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap="coolwarm", fmt=".2f") # Select only numeric columns for correlation calculation
plt.title("Correlation Heatmap")
plt.show()
```
