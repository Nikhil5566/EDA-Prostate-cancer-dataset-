```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Export.csv')
df.head(11)
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Year'] = df['Year'].fillna(df['Year'].mean())
```

```python
df.info()
```

```python
#  Changing the datatype of the Year columns:

df['Year'] = df['Year'].astype(float)
```

```python
df.info()
```

```python
df = df.rename(columns={'Domain Code': 'Domain_Code'})
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

# Visualization

```python
df['Area'].value_counts()
```

```python
# 1. Count plot: Distribution of products by country

plt.figure(figsize=(15, 8))
sns.countplot(data=df, x='Area', order=df['Area'].value_counts().index)
plt.title('Count of Products by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns

df[numeric_cols].hist(figsize=(14,12), bins=30)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
