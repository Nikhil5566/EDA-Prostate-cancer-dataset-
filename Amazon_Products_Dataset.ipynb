```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Amazon-Products.csv')
df.head(11)
```

```python
df['ratings'] = df['ratings'].fillna(df['ratings'].mode()[0])
df['no_of_ratings'] = df['no_of_ratings'].fillna(df['no_of_ratings'].mode()[0])
df['discount_price'] = df['discount_price'].fillna(df['discount_price'].mode()[0])
df['actual_price'] = df['actual_price'].fillna(df['actual_price'].mode()[0])
df.info()
df.isnull().sum()
```

```python
df.duplicated().sum()
```

# Univariant Analysis

```python
df['discount_price'].value_counts(normalize = True)
```

```python
# Histograms: For numerical variables
plt.figure(figsize = (15, 8))
plt.hist(df['discount_price'], bins=20)
plt.xlabel('discount_price')
plt.ylabel('Frequency')
plt.title('Distribution of Your Numerical Column')
plt.show()
```

```python
# Box plots:

plt.figure(figsize = (15, 8))
sns.boxplot(x=df['Unnamed: 0'])
plt.xlabel('Unnamed: 0')
plt.title('Box Plot of Your Numerical Column')
plt.show()
```

```python
# Count plots:

plt.figure(figsize = (15, 8))
sns.countplot(x=df['main_category'])
plt.xlabel('main_category')
plt.ylabel('Count')
plt.title('Count of Your Categorical Column')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels if they overlap
plt.show()
```

```python
# Summarize the 'ratings' column
print(df['ratings'].describe())

# Visualize the distribution with a histogram
plt.figure(figsize = (15, 8))
plt.hist(df['ratings'], bins=20)
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()

# Visualize the distribution with a box plot
plt.figure(figsize = (15, 8))
sns.boxplot(x=df['ratings'])
plt.xlabel('Ratings')
plt.title('Box Plot of Ratings')
plt.show()
```

# Bivariate Analysis

```python
# Example: Numerical vs. Numerical (e.g., 'discount_price' and 'no_of_ratings')
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['discount_price'], df['no_of_ratings'])
plt.xlabel('Discount Price')
plt.ylabel('Number of Ratings')
plt.title('Scatter Plot of Discount Price vs. Number of Ratings')
plt.show()
```

```python
# Example: Categorical vs. Numerical (e.g., 'main_category' and 'ratings')
# Box plot
plt.figure(figsize=(15, 8))
sns.boxplot(x='main_category', y='ratings', data=df)
plt.xlabel('Main Category')
plt.ylabel('Ratings')
plt.title('Box Plot of Ratings by Main Category')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(15, 8))
sns.violinplot(x='main_category', y='ratings', data=df)
plt.xlabel('Main Category')
plt.ylabel('Ratings')
plt.title('Violin Plot of Ratings by Main Category')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Example: Categorical vs. Categorical (e.g., 'main_category' and another categorical column if available)
# Assuming you have another categorical column, replace 'another_categorical_column'
# Count plot with hue
sns.countplot(x='main_category', hue='sub_category', data=df)
plt.xlabel('Main Category')
plt.ylabel('Count')
plt.title('Count of Main Category Grouped by Another Categorical Column')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Visualization

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='ratings', bins=20, kde=True) # kde=True adds a kernel density estimate curve
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.title('Distribution of Product Ratings')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.countplot(data=df, x='main_category', order=df['main_category'].value_counts().index)
plt.xlabel('Main Category')
plt.ylabel('Count')
plt.title('Number of Products per Main Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='discount_price', y='no_of_ratings', alpha=0.5) # alpha for transparency
plt.xlabel('Discount Price')
plt.ylabel('Number of Ratings')
plt.title('Discount Price vs. Number of Ratings')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=df, x='main_category', y='ratings')
plt.xlabel('Main Category')
plt.ylabel('Ratings')
plt.title('Ratings Distribution by Main Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(df.corr(numeric_only = True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
