# Handling Outliers

```python
Q1 = df['stress_level'].quantile(0.25)
Q3 = df['stress_level'].quantile(0.75)
IQR = Q3 - Q1

lower_bound  = Q1 - 1.5 * IQR
upper_bound = Q3 - 1.5 * IQR

# Create a new DataFrame by filtering out outliers
df_no_outliers = df[(df['stress_level'] >= lower_bound) & (df['stress_level'] <= upper_bound)].copy()

# Print the shape of the DataFrame after removing outliers
print("Shape of DataFrame after removing outliers:", df_no_outliers.shape)

# Display the first few rows of the original and modified DataFrames to see the difference
print("\nOriginal DataFrame (first 5 rows):")
display(df.head())

print("\nDataFrame after removing outliers (first 5 rows):")
display(df_no_outliers.head())
```

```python
# Visualization of Outliers:

plt.figure(figsize=(8, 6))
sns.boxplot(x=df['stress_level'])
plt.title('Box Plot of Stress Level with Outliers')
plt.xlabel('Stress Level')
plt.show()
```

# Univariate Analysis

```python
# For categorical columns
print("\nDescriptive statistics for categorical columns:")
display(df.describe(include='object'))
```

```python
# 2. Frequency Distribution (for Categorical Variables):

# Replace 'categorical_column' with the name of a categorical column
print("\nFrequency distribution for 'timestamp' column:")
display(df['timestamp'].value_counts())

# You can also get the percentage
print("\nPercentage distribution for 'timestamp' column:")
display(df['timestamp'].value_counts(normalize=True) * 100)
```

```python
# Replace 'numerical_column' with the name of a numerical column:

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='stress_level', kde=True)
plt.title('Distribution of Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box-Ploting:

plt.figure(figsize=(8, 6))
sns.boxplot(x=df['stress_level'])
plt.title('Box Plot of Stress Level')
plt.xlabel('Stress Level')
plt.show()
```

```python
# Replace 'numerical_column' with the name of a numerical column

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='stress_level', fill=True)
plt.title('Density Plot of Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Density')
plt.show()
```

```python
# Density Plots (KDE):

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='stress_level', fill=True)
plt.title('Density Plot of Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Density')
plt.show()
```

# Bivariate analysis

```python
# 1. Scatter Plots (Numerical vs. Numerical):

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='stress_level', y='crowd_density')
plt.title('Stress Level vs. Crowd Density')
plt.xlabel('Stress Level')
plt.ylabel('Crowd Density')
plt.show()
```

```python
# 2. Box Plots (Numerical vs. Categorical):

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='timestamp', y='stress_level')
plt.title('Stress Level by Timestamp')
plt.xlabel('Timestamp')
plt.ylabel('Stress Level')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 3. Correlation Matrix and Heatmap (Numerical vs. Numerical):

# Select only numerical columns
numerical_df = df.select_dtypes(include=np.number)

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
