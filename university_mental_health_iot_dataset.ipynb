# Visualization

```python
# 1. Histogram
# Replace 'numerical_column' with the name of a numerical column in your DataFrame
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='location_id', kde=True)
plt.title('Distribution of Numerical Column')
plt.xlabel('Numerical Column')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Countplot
# Replace 'categorical_column' with the name of a categorical column in your DataFrame
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='timestamp')
plt.title('Count of Categories in Categorical Column')
plt.xlabel('Categorical Column')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 3. Scatter plot
# Replace 'numerical_column_1' and 'numerical_column_2' with names of two numerical columns
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='stress_level', y='timestamp')
plt.title('Scatter plot between two numerical columns')
plt.xlabel('Numerical Column 1')
plt.ylabel('Numerical Column 2')
plt.show()
```

```python
# 4. Box plot
# Replace 'numerical_column' with a numerical column and 'categorical_column' with a categorical column
plt.figure(figsize=(10, 6))
# Correcting the column name for the y-axis to 'Mental_Stress_Status'
sns.boxplot(data=df, x='timestamp', y='crowd_density')
plt.title('Box plot of Numerical Column by Categorical Column')
plt.xlabel('Categorical Column')
plt.ylabel('Numerical Column')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 5. Correlation Heatmap
# Select only numerical columns for the correlation matrix
numerical_df = df.select_dtypes(include=np.number)
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
